```python
import pandas as pd
import numpy as np
import altair as alt
import eco_style
from io import StringIO
alt.themes.enable('light')
```

```python
with open("USTC_All_graph_coinmarketcap.csv") as f:
    csv = f.read()

csv = csv.replace('"\n"', '\n')
df = pd.read_csv(StringIO(csv), sep=";")

df['timestamp'] = df['timestamp'].str.replace('"', '', regex=False)
# e.g. 2025-03-17T00:00:00Z
df['timestamp'] = pd.to_datetime(df['timestamp'], format='mixed', errors='coerce')
df['timestamp'] = pd.to_datetime(df['timestamp'].dt.strftime('%Y-%m-%d'))

df = df.query("timestamp >= '2022-01-01' and timestamp <= '2022-12-31'")

line = alt.Chart(df).mark_line(
    color="#36B7B4"
).encode(
    x=alt.X('timestamp', title='', axis=alt.Axis(format='%b %Y')),
    y=alt.Y('close', title='', 
            axis=alt.Axis(format='$', title=''),
            scale=alt.Scale(type='linear')),
).properties(
    width=800,
    height=400
).interactive()

one_dollar_rule = alt.Chart(
    pd.DataFrame({'close': [1]})
).mark_rule(
    color="#676A86",
    strokeDash=[3, 3]
).encode(
    y='close',
    size=alt.value(1)
)

chart = line + one_dollar_rule
chart = chart.properties(
    width=750,
    height=350
)

chart.save("USTC_All_graph_coinmarketcap.png", scale_factor=3.0)
chart.save("USTC_All_graph_coinmarketcap.json")
chart
```

# USDT vs BTC

```python
tether_df = pd.read_csv("USDT_All_graph_coinmarketcap.csv", sep=";")
btc_df = pd.read_csv("BTC_All_graph_coinmarketcap.csv", sep=";")
tether_df['symbol'] = 'USDT'
btc_df['symbol'] = 'BTC'
df = pd.concat([tether_df, btc_df])
df = df[['close', 'timestamp', 'symbol']]
df['timestamp'] = df['timestamp'].str.replace('"', '', regex=False)
df['timestamp'] = pd.to_datetime(df['timestamp'], format='mixed', errors='coerce')
df['timestamp'] = pd.to_datetime(df['timestamp'].dt.strftime('%Y-%m-%d'))

# what's the value for each at the start of 2024?
USDT_val = df.query("timestamp == '2024-01-01' and symbol == 'USDT'")['close'].values[0]
BTC_val = df.query("timestamp == '2024-01-01' and symbol == 'BTC'")['close'].values[0]

df.loc[df['symbol'] == 'USDT', 'close'] = df.loc[df['symbol'] == 'USDT', 'close'] / USDT_val
df.loc[df['symbol'] == 'BTC', 'close'] = df.loc[df['symbol'] == 'BTC', 'close'] / BTC_val

df['close'] = df['close']*100

df = df.query("timestamp >= '2017-01-01' ")

# end labels
df['label']  = np.where(df['timestamp'] == '2025-03-01', df['symbol'], '')

lines = alt.Chart(df).mark_line().encode(
    x=alt.X('timestamp', title='', axis=alt.Axis(format='%b %Y')),
    y=alt.Y('close', title='', 
            axis=alt.Axis(
                labelExpr="(datum.value == 100 ? 'Jan 2024=' : '')+datum.label",
                title=''),
            scale=alt.Scale(type='linear')),
    color=alt.Color('symbol', legend=None)
).properties(
    width=800,
    height=400
)

end_labels = lines.mark_text(align='left', dx=5, dy=-5).encode(
    text='label'
)

chart = lines + end_labels

chart.save("USTC_BTC_USDT_All_graph_coinmarketcap.png", scale_factor=3.0)
chart.save("USTC_BTC_USDT_All_graph_coinmarketcap.json")
chart
```
