```python
import pandas as pd
import requests
```

```python
states = ["AZ", "GA", "MI", "NV", "NC", "PA", "WI"]
```

# Unemp Data

```python
data = []
base_prefix = ""
base_suffix = "UR"

for state in states:
    url = f"https://api.stlouisfed.org/fred/series/observations?series_id={base_prefix}{state}{base_suffix}&api_key=22ee7a76e736e32f54f5df0a7171538d&file_type=json"
    print(url)
    response = requests.get(url)
    data = response.json()['observations']
    # add the state to the data
    for d in data:
        d['state'] = state
    data.extend(data)

unemp_df = pd.DataFrame(data)
unemp_df
```
