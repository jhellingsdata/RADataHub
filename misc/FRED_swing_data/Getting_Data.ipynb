```python
import requests
import pandas as pd

states = ["AZR", "GAR", "MIR", "NVR", "NCR", "PAR", "WIR"]
all_data = []  # To store all states' data
base_suffix = "PIPC"
api_key = "22ee7a76e736e32f54f5df0a7171538d"

for state in states:
    url = f"https://api.stlouisfed.org/fred/series/observations?series_id={state}{base_suffix}&api_key={api_key}&file_type=json"
    response = requests.get(url)
    state_data = response.json().get('observations', [])
    
    # Add the state identifier to each observation
    for d in state_data:
        d['state'] = state
    
    # Append the state's data to the overall list
    all_data.extend(state_data)

# Convert to DataFrame
inc_data = pd.DataFrame(all_data)
inc_data
```

```python
# Define the mapping of state codes to state names
state_names = {
    "AZR": "Arizona",
    "GAR": "Georgia",
    "MIR": "Michigan",
    "NVR": "Nevada",
    "NCR": "North Carolina",
    "PAR": "Pennsylvania",
    "WIR": "Wisconsin"
}

# Add a new 'name' column to inc_data by mapping the 'state' column
inc_data['name'] = inc_data['state'].map(state_names)
inc_data
```

```python
# Ensure the 'date' column is in datetime format
inc_data['date'] = pd.to_datetime(inc_data['date'])

# Filter for rows where the year in 'date' is between 2020 and 2024
inc_data = inc_data[(inc_data['date'].dt.year >= 2018) & (inc_data['date'].dt.year <= 2024)]
inc_data
```

```python
import altair as alt
import pandas as pd

# Sample DataFrame assuming `date`, `value`, and `stname` columns
# Filtered DataFrame for Jan 2020 and Jan 2024 only
df_slope = inc_data[inc_data['date'].isin(['2018-01-01', '2022-01-01'])]

# Convert dates to categorical labels
df_slope['date_label'] = df_slope['date'].apply(lambda x: 'Jan 2018' if x == '2018-01-01' else 'Jan 2022')

# Slope Graph
slope_chart = alt.Chart(df_slope).mark_line(point=True).encode(
    x=alt.X('date_label:N', title='', scale=alt.Scale(padding=0)),
    y=alt.Y('value:Q', title='',
            axis=alt.Axis(labelExpr="'$' + format(datum.value,',')")),
    color=alt.Color('name:N', legend=None),
    tooltip=['name:N', 'date_label:N', 'value:Q']
).properties(
    width=500,
    height=300,
    title={
        "text": "USA swing states: Real Per Capita Personal Income",
        "anchor": "start",
        "subtitle": ["Chained 2017 USD", "Source: BLS 2024", ""],
        "subtitleColor": "#676A86"
    }
)

# Adding Text Labels only at 2024
text_labels = alt.Chart(df_slope[df_slope['date_label'] == 'Jan 2022']).mark_text(
    align='left',
    dx=5,
    dy=2
    # {
    #     "expr": "datum.stname == 'Pennsylvania' ? 4 : datum.stname == 'North Carolina' ? -8 : datum.stname == 'Arizona' ? 0 : datum.stname == 'Georgia' ? -6 : 0"
    # }
).encode(
    x=alt.X('date_label:N'),
    y=alt.Y('value:Q'),
    text='name:N',
    color=alt.Color('name:N', legend=None)
)

# Combine line and text
slope_graph = slope_chart + text_labels
slope_graph
```
