```python
import pandas as pd
```

## Chinese Coal 

```python
df = pd.read_json("https://api.economicsobservatory.com/chn/coal?vega")
df['value'] = df['value'] * 10_000

df['year'] = df.date.astype(str).str[:4]

df = df[['year', 'value']]

df.to_csv("/Users/finn/Documents/GitHub/courses/courses/1/example_data/chinese_coal_consumption.csv", index=False)

df.head()
```

## GDP PC vs Life Expectency

```python
df = pd.read_csv("UN_living_standards.csv")

df = df[['Country', 'GDP per capita (PPP)', 'Life Expectancy']]

df["Country"] = df.Country.str.strip("*")

df = df.rename(columns={"GDP per capita (PPP)": "GDP per-capita"})

df.to_csv("/Users/finn/Documents/GitHub/courses/courses/1/example_data/gdp_and_life_expectency.csv", index=False)

df.head()
```

#### Debt

```python
df = pd.read_csv("https://raw.githubusercontent.com/jhellingsdata/RADataHub/main/Debt/data/nominal_debt_2023_with_ids.csv")

df = df[['country', "iso3", 'debt_gdp']]

p21_iso3s = ["BGD", "BRA", "CHN", "COD", "EGY", "ETH", "GBR", "DEU", "IND", "IDN", "IRN", "JPN", "MEX", "NGA", "PAK", "PHL", "RUS", "THA", "TUR", "USA", "VNM"]

df = df[df.iso3.isin(p21_iso3s)]

df = df.rename(columns={"country": "Country", "debt_gdp": "Debt Ratio"})

df = df[['Country', 'Debt Ratio']]

df.to_csv("/Users/finn/Documents/GitHub/courses/courses/1/example_data/gov_debt_gdp_ratio.csv", index=False)
```
