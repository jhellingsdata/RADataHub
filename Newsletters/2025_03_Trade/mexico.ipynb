```python
import pandas as pd
import altair as alt
import requests
import eco_style
import json
alt.themes.enable("light")
```

```python
image = alt.Chart(pd.DataFrame([{'url': 'https://raw.githubusercontent.com/EconomicsObservatory/ECOvisualisations/refs/heads/main/guidelines/logos/eco-logo-light.png'}])).mark_image(
    height=30,
    width=90,
    opacity=0.5,
    # align='right',
    baseline='bottom'
).encode(
    url='url',
    x=alt.value(400),
    y=alt.value(0)
)
```

# Chart: Mexican trade partners

```python
df
```

```python
df = pd.read_csv("data/mexico_export_partners.csv")
df = df.reset_index().rename(columns={'index': 'order'})
df['value'] = df['value'].astype(float) / 100


bars = alt.Chart(df).mark_bar(
    color="#36B7B4"
).encode(
    y=alt.Y('partner:N', sort=alt.EncodingSortField(field='order', order='ascending'), title=''),
    x=alt.X('value:Q', title='', axis=alt.Axis(format='~%')),
)

lables = bars.mark_text(
    align='left',
    baseline='middle',
    dx=5,
    fontSize=12,
    color='#676A86'
).encode(
    text=alt.Text('value:Q', format='.0%'),
    x=alt.X('value:Q', title='', axis=alt.Axis(format='~%')),
    y=alt.Y('partner:N', sort=alt.EncodingSortField(field='order', order='ascending'), title=''),
)

chart = bars + lables + image.encode(
    x=alt.value(350),
    y=alt.value(-10)
).properties(
    width=400,
    height=270,
)

chart.save('charts/mexico_export_partners.png', scale_factor=2)
chart.save('charts/mexico_export_partners.json')
chart
```

# What does Mexico export to the US?


```python
with open("data/mexico_exports_to_us.json") as f:
    data = json.load(f)
df = pd.DataFrame(data['data'])
df.query("`Section ID` == 1")

df = df.groupby(by=['Section ID', 'Section']).agg(
    {'Trade Value': 'sum'}
).reset_index()

df['Trade Value'] = df['Trade Value'].astype(float) / 1000000000

chart = alt.Chart(df.head(15)).mark_bar(
    color="#36B7B4"
).encode(
    y=alt.Y('Section:N', sort=alt.EncodingSortField(field='Trade Value', order='descending'), title=''),
    x=alt.X('Trade Value:Q', title='', axis=alt.Axis(
        format='$', labelExpr="datum.label + 'B' "
    )),
)

chart = (chart + image.encode(
    x=alt.value(325),
    y=alt.value(-10)
)).properties(
    width=400,
    height=350
)

chart.save('charts/mexico_exports_to_us.png', scale_factor=2)
chart.save('charts/mexico_exports_to_us.json')
chart
```
