## Vacancies

```python
import pandas as pd
```

```python
vacs = pd.read_csv('vacs02feb2026.csv')

# drop na from Date col
vacs = vacs.dropna(subset=['Date'])

# make date datetime from "Apr-Jun 2001" to "2001-06-01"
vacs['Date'] = (
    vacs['Date']
    .str.strip()  # Remove leading/trailing whitespace
    .str.replace(r'.*-\s*(\w{3})\s+(\d{4})', r'\1 \2', regex=True)
    .pipe(pd.to_datetime, format='%b %Y')
)

# melt
vacs = vacs.melt(id_vars=['Date'], var_name='Sector', value_name='Vacancies')

# drop total services
vacs = vacs[vacs['Sector'] != 'Total services ']

# remove "- " from Sector
vacs['Sector'] = vacs['Sector'].str.replace('- ', '', regex=True)

# remove dashes from Sector
vacs['Sector'] = vacs['Sector'].str.replace('-', '', regex=True)

# remove numeric characters from Sector
vacs['Sector'] = vacs['Sector'].str.replace(r'\d+', '', regex=True)

# remove commas from the END of Sector
vacs['Sector'] = vacs['Sector'].str.replace(r',\s*$', '', regex=True)

vacs.to_csv('vacs.csv', index=False)
```
