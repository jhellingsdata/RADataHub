```python
import requests
import pandas as pd

# OECD member ISO3 codes (38 members as of 2025)
OECD = [
    "AUS", "AUT", "BEL", "CAN", "CHL", "COL", "CRI", "CZE", "DNK", "EST",
    "FIN", "FRA", "DEU", "GRC", "HUN", "ISL", "IRL", "ISR", "ITA", "JPN",
    "KOR", "LVA", "LTU", "LUX", "MEX", "NLD", "NZL", "NOR", "POL", "PRT",
    "SVK", "SVN", "ESP", "SWE", "CHE", "TUR", "GBR", "USA",
]

# World Bank API: GFCF as % of GDP, most recent non-empty value
url = "https://api.worldbank.org/v2/country/{}/indicator/NE.GDI.FTOT.ZS"
params = {"format": "json", "per_page": 300, "mrnev": 1}

# You can request all at once by joining codes with ";"
codes = ";".join(OECD)
r = requests.get(url.format(codes), params=params)
data = r.json()[1]

rows = [
    {
        "country": d["country"]["value"],
        "iso3": d["countryiso3code"],
        "gfcf_pct_gdp": d["value"],
        "year": int(d["date"]),
    }
    for d in data
    if d["value"] is not None
]

df = pd.DataFrame(rows).sort_values("gfcf_pct_gdp", ascending=False).reset_index(drop=True)
print(df.to_string())
```

```python
# save to csv
df.to_csv("gfcf_pct_gdp_oecd.csv", index=False)
```

```python
df = pd.read_excel(
    "grossfixedcapitalformationbysectorandasset.xlsx",
    sheet_name="G1_CVM_SA_Q_levels",
    header=None,
    skiprows=7,
)

df = df.iloc[:, [0, 1, 2, 3, 4, 5, 6, 12]]
df.columns = [
    "period",
    "Business Investment (excl. dwellings)",
    "General Government",
    "public_corp_dwellings",
    "public_corp_transfer",
    "private_sector_dwellings",
    "private_sector_transfer",
    "total_gfcf",
]

df = df.dropna(subset=["period"])
df = df[df["period"].astype(str).str.match(r"^\d{4} Q[1-4]$")]

num_cols = df.columns.drop("period")
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

# Combine sub-columns
df["Public Corporations (housing)"] = df["public_corp_dwellings"] + df["public_corp_transfer"]
df["Private Housing & Transfer Costs"] = df["private_sector_dwellings"] + df["private_sector_transfer"]

# Parse period to datetime (start of quarter)
df["period"] = pd.PeriodIndex(df["period"].str.replace(" ", ""), freq="Q").to_timestamp()

# Select and melt
sector_cols = [
    "Business Investment (excl. dwellings)",
    "General Government",
    "Public Corporations (housing)",
    "Private Housing & Transfer Costs",
]

melted = (
    df[["period"] + sector_cols]
    .reset_index(drop=True)
    .melt(id_vars="period", var_name="Sector", value_name="Value")
)

# make Value the absolute value of itself
melted["Value"] = melted["Value"].abs()


melted.head()
```

```python
melted.to_csv("gfcf_by_sector.csv", index=False)
```

```python
"""
UK vs US Scale-Up Finance: Data Pull for Vega-Lite
====================================================
Output:
  uk_us_nfc_finance_long.csv — long-form for a faceted line chart
  Columns: date, country, indicator, value

Requirements:
  pip install fredapi pandas --break-system-packages

FRED API key: https://fred.stlouisfed.org/docs/api/api_key.html
"""

import os
import pandas as pd
from fredapi import Fred

# ── CONFIG ──────────────────────────────────────────────────────────────────
FRED_API_KEY = os.environ.get("FRED_API_KEY", key)
fred = Fred(api_key=FRED_API_KEY)
START = "1990-01-01"

# ── FRED SERIES ─────────────────────────────────────────────────────────────
#
# Panel 1: NFC debt securities outstanding (BIS Debt Securities Statistics)
#   UNITS: Millions of USD.  ← NB: millions, not billions
#   https://data.bis.org/topics/DSS
#
# Panel 2: Total credit to NFCs, % of GDP (BIS credit-to-NFS long series)
#   Break-adjusted. Methodology: Dembiermont et al. (2013),
#   BIS Quarterly Review, March 2013.
#   https://www.bis.org/publ/qtrpdf/r_qt1303h.htm
#
# Panel 3 (derived): NFC debt securities as share of total NFC credit
#   = bonds outstanding (millions USD, converted to billions)
#     / total NFC credit (billions USD) × 100
#
#   CRITICAL: TDSAMRIAONCUS/GB are in MILLIONS of USD,
#   while CRDQUSANABIS/QGBNAMUSDA are in BILLIONS of USD.
#   Must divide bonds by 1000 before computing the ratio.

series_map = {
    # Panel 1: NFC debt securities outstanding
    # Units: Millions of USD. Quarterly, NSA. Ends Q4 2022.
    "TDSAMRIAONCUS": {"country": "US", "indicator": "NFC debt securities outstanding (USD bn)", "unit_factor": 1000},
    "TDSAMRIAONCGB": {"country": "UK", "indicator": "NFC debt securities outstanding (USD bn)", "unit_factor": 1000},

    # Panel 2: Total credit to NFCs (% of GDP)
    "QUSNAM770A": {"country": "US", "indicator": "Total credit to NFCs (% of GDP)", "unit_factor": 1},
    "QGBNAM770A": {"country": "UK", "indicator": "Total credit to NFCs (% of GDP)", "unit_factor": 1},

    # Panel 3 inputs: Total credit to NFCs in USD (billions)
    "CRDQUSANABIS": {"country": "US", "indicator": "_total_nfc_credit_usd_bn", "unit_factor": 1},
    "QGBNAMUSDA":   {"country": "UK", "indicator": "_total_nfc_credit_usd_bn", "unit_factor": 1},
}

# ── FETCH DATA ──────────────────────────────────────────────────────────────
raw = {}
for series_id, meta in series_map.items():
    try:
        s = fred.get_series(series_id, observation_start=START)
        s.name = series_id
        raw[series_id] = s
        print(f"  ✓ {meta['country']} | {meta['indicator']} ({series_id}): "
              f"{len(s)} obs, {s.index.min().date()} → {s.index.max().date()}")
    except Exception as e:
        print(f"  ✗ {series_id}: {e}")

# ── BUILD LONG-FORM DATAFRAME ──────────────────────────────────────────────
records = []

# Panels 1 & 2: direct series (skip internal helpers starting with _)
for series_id, meta in series_map.items():
    if series_id not in raw:
        continue
    if meta["indicator"].startswith("_"):
        continue
    s = raw[series_id].dropna()
    factor = meta["unit_factor"]
    for date, value in s.items():
        records.append({
            "date": date.strftime("%Y-%m-%d"),
            "country": meta["country"],
            "indicator": meta["indicator"],
            "value": round(float(value) / factor, 2),
        })

# Panel 3: Bond share of total NFC credit (derived)
# bonds are in MILLIONS, total credit in BILLIONS → divide bonds by 1000
panel3_pairs = [
    ("US", "TDSAMRIAONCUS", "CRDQUSANABIS"),
    ("UK", "TDSAMRIAONCGB", "QGBNAMUSDA"),
]

for country, bonds_id, total_id in panel3_pairs:
    if bonds_id not in raw or total_id not in raw:
        print(f"  ⚠ Cannot compute bond share for {country} — missing series")
        continue
    bonds = raw[bonds_id].dropna() / 1000  # millions → billions
    total = raw[total_id].dropna()
    common = bonds.index.intersection(total.index)
    if len(common) == 0:
        print(f"  ⚠ No overlapping dates for {country} bond share")
        continue
    bonds = bonds.loc[common]
    total = total.loc[common]
    share = (bonds / total) * 100
    for date, value in share.items():
        if pd.notna(value) and value >= 0:
            records.append({
                "date": date.strftime("%Y-%m-%d"),
                "country": country,
                "indicator": "NFC bonds as share of total NFC credit (%)",
                "value": round(float(value), 2),
            })

df_long = pd.DataFrame(records)
df_long = df_long.sort_values(["indicator", "country", "date"]).reset_index(drop=True)

# ── SAVE ────────────────────────────────────────────────────────────────────

df_long.to_csv(f"uk_us_nfc_finance_long.csv", index=False)
print(f"\nSaved: uk_us_nfc_finance_long.csv")

# ── SUMMARY ─────────────────────────────────────────────────────────────────
print(f"\nLong-form dataframe: {len(df_long)} rows")
print(f"Indicators: {df_long['indicator'].unique().tolist()}")
print(f"Date range: {df_long['date'].min()} → {df_long['date'].max()}")
print("\nIndicators (use as Vega-Lite facet field):")
for ind in df_long["indicator"].unique():
    subset = df_long[df_long["indicator"] == ind]
    countries = subset["country"].unique().tolist()
    print(f"  • {ind}")
    for c in countries:
        cs = subset[subset["country"] == c]
        print(f"    {c}: {cs['date'].min()} → {cs['date'].max()}, "
              f"{len(cs)} obs, latest = {cs['value'].iloc[-1]}")
```
