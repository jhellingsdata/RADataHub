```python
import pandas as pd
```

```python
# change from "YYYY QX" to quarter start timestamp
gfcf['Time Period'] = pd.PeriodIndex(
    gfcf['Time Period'].str.replace(' Q', 'Q'),
    freq='Q'
).to_timestamp(how='start')

# remove [] and everything inside them AND the space before it
gfcf.columns = gfcf.columns.str.replace(r'\s*\[.*\]', '', regex=True)

gfcf
```

```python
# remove "Total sector (S.1) " from col names
gfcf.columns = gfcf.columns.str.replace('Total sector \(S\.1\) ', '', regex=True)

gfcf
```

```python
gfcf.drop(
    columns=['Total asset (GFCF)'],
    inplace=True
)

gfcf
```

```python
gfcf = gfcf.melt(id_vars='Time Period', var_name='Asset Type', value_name='Value')

gfcf.to_csv('gfcf_components.csv', index=False)
```
