```python
df = pd.read_csv("maddison_gdp.csv")
df = df.query("Code == 'KOR' | Code == 'PRK'")
df = df.iloc[:, :-1]
df = df.query("Year >= 1940")
df.columns = ['country', 'id', 'date', 'value']

df['label']  = np.where(df['date']==2022, df.country, '')

df['date'] = pd.to_datetime(df['date'], format='%Y')

base = alt.Chart(df).encode(
    x=alt.X('date:T', title=''),
    y=alt.Y('value:Q', title='', axis={
        "labelExpr": "'$'+ datum.label+(datum.value == 45000 ? ' pc (2011 PPP)' : '')"
    }),
    color=alt.Color('country:N', title='Country', legend=None)
)

lines = base.mark_line().encode(
    detail='id:N'
)

labels = base.mark_text(align='left', dx=5).encode(
    text='label',
    opacity=alt.OpacityValue(0.8)
)

chart = lines + labels

chart = chart.properties(
    width=500,
    height=400,
    background='rgba(247, 247, 247, 1)',
    title={
        "text": " ",
        "subtitle": "GDP per capita (2011 PPP) of North and South Korea, Source: Maddison Project Database",
        "subtitleColor": "rgb(0,0,0,0.4)",
        "align": "left",
        "anchor": "start",
        "frame": "group",
    }
)

chart.save("kor_gdp.json")
chart.save("kor_gdp.png", scale_factor=3)
chart
```
