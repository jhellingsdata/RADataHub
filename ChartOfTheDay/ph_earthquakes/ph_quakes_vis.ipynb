```python
from ecostyles import EcoStyles
# Create styles instance
styles = EcoStyles()

import altair as alt
import pandas as pd
```

```python
# Register and enable a theme
styles.register_and_enable_theme(theme_name="article")
```

```python
# Load csv files
quake_df = pd.read_csv('ph_earthquakes.csv')

# Reformat date and extract year
quake_df['time'] = pd.to_datetime(quake_df['time'])
quake_df['year'] = quake_df['time'].dt.year

# Filter to magnitude 6.0+
quake_df = quake_df[quake_df['mag'] >= 6.0]

# Add flag for most recent earthquake

quake_df['is_main'] = quake_df['mag'] == quake_df['mag'].max()
quake_df
```

```python
# Plot magnitude of earthquakes over time, with most recent highlighted
quake_scatter = alt.Chart(quake_df).mark_circle().encode(
    x = alt.X(
        'time:T', 
        title=None, 
        scale=alt.Scale(domain=['2021-01-01', '2026-12-31']),
        axis=alt.Axis(format='%Y', tickCount='year', grid=False)),
    y = alt.Y(
        'mag:Q', 
        title='Magnitude', 
        scale=alt.Scale(domain=[5.9, 8]),
        axis=alt.Axis(grid=False)),
    size = alt.Size(
        'mag:Q', 
        scale=alt.Scale(domain=[6.0, 7.8], range=[50, 500]), 
        legend=None),
    color=alt.Color(
        'mag:Q',
        scale=alt.Scale(scheme='orangered'),
        legend=None
    )
)

halo = alt.Chart(quake_df[quake_df['is_main']]).mark_circle(opacity=0.2).encode(
    x=alt.X('time:T'), y=alt.Y('mag:Q'),
    size=alt.value(1200), color=alt.value('#d62728')
)
label = alt.Chart(quake_df[quake_df['is_main']]).mark_text(
    dy=-25, fontSize=11, fontWeight='bold'
).encode(
    x=alt.X('time:T'), y=alt.Y('mag:Q'),
    text=alt.value('Mw 7.8')
)

quake_scatter = alt.layer(quake_scatter, halo, label)
quake_scatter
```

```python
# Save charts
styles.save(quake_scatter, 'visualisation', 'ph_quakes_scatter', width=450, height=360)
```
