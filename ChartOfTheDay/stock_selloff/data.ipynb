```python
import matplotlib.pyplot as plt

# Create a copy of closes_long for indexing
indexed_closes = closes_long.copy()

# Group by Index (country) and find the value on January 21, 2025 for each country
base_date = '2025-01-20'
base_values = indexed_closes[indexed_closes['Date'] == base_date].set_index('Index')['Close']

# Add a column for indexed values (Jan 21, 2025 = 100)
def index_values(group):
    country = group['Index'].iloc[0]
    if country in base_values:
        base_value = base_values[country]
        if not pd.isna(base_value) and base_value != 0:
            group['Indexed_Close'] = (group['Close'] / base_value) * 100
    return group

# Apply the indexing function to each country group
indexed_closes = indexed_closes.groupby('Index', group_keys=False).apply(index_values)

# show indexed values
indexed_closes
```
