```python
data = {
    "Academic Year": [
        "2006-07", "2007-08", "2008-09", "2009-10", "2010-11",
        "2011-12", "2012-13", "2013-14", "2014-15", "2015-16",
        "2016-17", "2017-18", "2018-19", "2019-20", "2020-21",
        "2021-22", "2022-23", "2023-24", "2024-25"
    ],
    "Number of International Students": [
        3941, 3961, 4025, 4201, 4207, 4347, 4531, 4633, 4822, 4961,
        5245, 5483, 5542, 5718, 5359, 5473, 6527, 6713, 6793
    ],
    "International % of Total Enrollment": [
        19.6, 19.8, 19.8, 19.9, 19.7, 20.6, 21.3, 21.8, 22.5, 22.7,
        23.4, 23.9, 23.8, 24.1, 24.1, 25.8, 25.8, 26.8, 27.2
    ]
}


df = pd.DataFrame(data)
df["International % of Total Enrollment"] = df["International % of Total Enrollment"] / 100
```

https://oneworld.worldwide.harvard.edu/international-students-at-harvard/

```python
bars = alt.Chart(df).mark_bar(color="#36B7B4").encode(
    x=alt.X("Academic Year:N", title=None, axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("Number of International Students:Q", axis=alt.Axis(title="Number of International Students"))
)

line = alt.Chart(df).mark_line(color="#E6224B", point=alt.OverlayMarkDef(color="#E6224B")).encode(
    x="Academic Year:N",
    y=alt.Y(
        "International % of Total Enrollment:Q",
        axis=alt.Axis(
            title="Share enrolled",
            format=".0%"  # Format as percentage
        ),
        scale=alt.Scale(domain=[0.15, 0.30])
    )
)

# Add text labels showing percentage values at each point
text = alt.Chart(df).mark_text(
    align='center',
    baseline='bottom',
    dy=-5,
    fontSize=10,
    color="#E6224B"
).encode(
    x="Academic Year:N",
    y=alt.Y("International % of Total Enrollment:Q", 
            scale=alt.Scale(domain=[0.15, 0.30]),
            axis=None),  # Remove axis to prevent duplicate title
    text=alt.Text("International % of Total Enrollment:Q", format=".0%")
)

chart = alt.layer(bars, line, text).resolve_scale(
    y='independent'
).properties(
    width=700,
    height=400,
    title={
        "text": ["International Students at Harvard"],
        "subtitle": ["Number and share of international students at Harvard university", "Source: Harvard University", ""],
        "subtitleColor": "#676A86",
        "anchor": "start",
        "fontSize": 22 
    }
)

chart
```
