```python
oecd_housing = pd.read_excel('HM_1_1_OECD.xlsx', sheet_name="Sheet1")
```

```python
oecd_housing.columns
```

```python
oecd_housing = oecd_housing.sort_values(by='percentage vacant dwellings and seasonal or holiday homes', ascending=False)
oecd_housing.reset_index(drop=True, inplace=True)
oecd_housing['percentage vacant dwellings excluding seasonal and holiday homes'] = oecd_housing['percentage vacant dwellings excluding seasonal and holiday homes'] / 100
oecd_housing['percentage seasonal or holiday homes'] = oecd_housing['percentage seasonal or holiday homes'] / 100
oecd_housing['percentage vacant dwellings and seasonal or holiday homes'] = oecd_housing['percentage vacant dwellings and seasonal or holiday homes'] / 100
#oecd_housing
```

```python
chart = alt.Chart(oecd_housing).mark_bar().encode(
    x=alt.X(
        'Country:O', 
        title=None, 
        axis=alt.Axis(labelAngle=270),
        sort=alt.EncodingSortField(
            field='percentage vacant dwellings and seasonal or holiday homes',  
            order='descending'  
        )
    ),
    y=alt.Y(
        'percentage vacant dwellings and seasonal or holiday homes:Q', 
        axis=alt.Axis(title='% Vacant dwellings and seasonal or holiday homes',  format='%')
    ),
    color=alt.condition(
        alt.datum.Country == 'UK (England)',  
        alt.value('#E6224B'),
        alt.value('#122B39')
    ),
)

chart = chart.properties(
    title={
        "text": ["Vacant housing across the OECD"],
        "subtitle": ["Percentage of vacant dwellings and seasonal/holiday homes as a share of the total dwelling stock, 2022 or latest year available", "Source: OECD", ""],
        "subtitleColor": "#676A86",
        "anchor": "start"
    }
)

chart
```

```python
chart.save('dwellings_oecd.json')
```

```python
rent = pd.read_excel('priceindexrent.xlsx', sheet_name="Table 1")
#rent
```

```python
df_z = rent[rent['Region or country name'].str.contains('\[z\]')]
df_z = df_z[['Time period', 'Area name',  'Annual change',]]
df_z = df_z[df_z['Annual change'] != '[x]']
df_z['Annual change'] = pd.to_numeric(df_z['Annual change'])
df_z['Time period'] = pd.to_datetime(df_z['Time period'])
df_z['Annual change'] = df_z['Annual change'] / 100
df_z = df_z.rename(columns={'Annual change': 'Percentage change'})
```

```python
df = df_z[df_z['Area name'].isin(['UK', 'London'])]
#df
```

```python
df_land = df_z[df_z['Area name'].isin(['England', 'London', 'Wales', 'Scotland', 'Northern Ireland'])]
df_UK = df_z[df_z['Area name'].isin(['UK'])]
```

```python
last_values = df_land.sort_values('Time period').groupby('Area name').last().reset_index()
last_values

last_values_UK = df_UK.sort_values('Time period').groupby('Area name').last().reset_index()
last_values_UK
```

```python
#Private rental price percentage change over 12 months, UK and London, January 2016 to January 2024
line_chart = alt.Chart(df_land).mark_line(strokeWidth=1).encode(
    x=alt.X('Time period:T'),
    y=alt.Y('Percentage change:Q', axis=alt.Axis(format='%')),
    color=alt.Color('Area name:O',  scale=alt.Scale(range=["#00A767","#179FDB","#0063AF","#122B39","#36B7B4"]), legend=None),
    tooltip=['Time period', 'Area name', 'Percentage change']
).properties(
    width=500,
    height=400
).transform_filter(
    alt.datum['Area name'] != 'UK'  
)

line_chart = line_chart.properties(
    title={
        "text": ["Trends in UK private housing rental prices"],
        "subtitle": ["Private rental price index percentage change over 12 months, United Kingdom, 2016 - 2024 ", "Source: ONS", ""],
        "subtitleColor": "#676A86",
        "anchor": "start"
    }
)

# 
text = alt.Chart(last_values).mark_text(align='left', dx=5, dy=-2).encode(
    x='Time period:T',
    y='Percentage change:Q',
    text='Area name:O',
    color='Area name:O'
).transform_filter(
    alt.datum['Area name'] != 'UK' )

text2 = alt.Chart(last_values_UK).mark_text(align='left', dx=-5, dy=-15).encode(
    x='Time period:T',
    y='Percentage change:Q',
    text='Area name:O',
    color=alt.value("#E6224B"),  
)


uk_line = alt.Chart(df_UK).mark_line(strokeDash=[1, 1], strokeWidth=1.5).encode(
    x=alt.X('Time period:T'),
    y=alt.Y('Percentage change:Q', axis=alt.Axis(format='%')),
    color=alt.value("#E6224B"),  
    tooltip=['Time period', 'Area name', 'Percentage change']
).transform_filter(
    alt.datum['Area name'] == 'UK'  
)

final_chart = line_chart + uk_line + text + text2
final_chart
```

```python
chart.save('private_rent_UK.json')
```
