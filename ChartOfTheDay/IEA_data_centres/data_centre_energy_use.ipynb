```python
import pandas as pd
import altair as alt
```

```python
from ecostyles import EcoStyles
# Create styles instance
styles = EcoStyles()
# Register and enable a theme
styles.register_and_enable_theme(theme_name="article")  # or "article"
```

```python
IEA = pd.read_excel('Data_annex_Energy_and_AI.xlsx', sheet_name='World Data', skiprows=2)
IEA = IEA.iloc[20:27, 2:8]
IEA.columns = ['Usage (TWh)', '2020', '2023', '2024', 'drop', '2030 IEA Base Projection']
IEA = IEA.drop(columns=['2020', '2023', 'drop'])
IEA = IEA.melt(id_vars='Usage (TWh)', var_name='Year', value_name='Energy Use (TWh)')
IEA['Usage (TWh)'] = IEA['Usage (TWh)'].replace('Colocation and service provider', 'Colocation and\nservice provider')
IEA_chart_data = IEA[~IEA['Usage (TWh)'].isin(['IT', 'Total'])]
```

```python
iea_chart = alt.Chart(IEA_chart_data).mark_bar().encode(
    x=alt.X('Usage (TWh)', axis=alt.Axis(labelFontSize=14, labelLimit=200, labelExpr="split(datum.value, '\\n')")),
    xOffset='Year:N',
    y=alt.Y('Energy Use (TWh)', axis=alt.Axis(title='Energy Use (TWh)', titleFontSize=14, labelFontSize=14)),
    color=alt.Color('Year:N', legend=alt.Legend(title="Year", titleFontSize=14, labelFontSize=14, orient='bottom', labelLimit=300))
).properties(
    title=alt.TitleParams(
        text='IEA Projected Data Centre Energy Use',
        subtitle='Source: IEA. License: CC BY 4.0',
        subtitleFontSize=11,
        anchor='start'
    ),
    height=350,
    width=600
).configure_title(fontSize=16, anchor='start')

iea_chart
```

```python
# Save to png
iea_chart.save('iea_projected_data_centre_energy_use.png', scale_factor=2)
# Save to json
iea_chart.save('iea_projected_data_centre_energy_use.json', scale_factor=2)
```

```python
IEA_total = IEA[IEA['Usage (TWh)'] == 'Total']
total_2024 = IEA_total[IEA_total['Year'] == '2024']['Energy Use (TWh)'].values[0]
total_2030 = IEA_total[IEA_total['Year'] == '2030 IEA Base Projection']['Energy Use (TWh)'].values[0]

pct_increase = (total_2030 - total_2024) / total_2024 * 100

print(f"Projected percentage increase in total data centre energy use from 2024 to 2030: {pct_increase:.2f}%")
```
