```python
groceries_list = {
    210111: "Bread",
    211709: "Milk",
    211306: "Butter",
    210213: "Cereal",
    210910: "Chicken",
    211101: "Fish",
    212719: "Bananas",
    212211: "Chocolate"
}
```

```python
items_df[items_df['item_id'].isin(groceries_list.keys())]
```

```python
full_df = pd.merge(prices_df, items_df[['item_id', 'description']], on='item_id')
```

```python
full_df['quote_date'] = pd.to_datetime(full_df['quote_date'], format='%Y%m')
```

```python
groceries_df = full_df[full_df['item_id'].isin(groceries_list.keys())]

groceries_df
```

```python
shopping_list = groceries_df.groupby(['quote_date', 'description', 'item_id']).agg({'price': 'mean'}).reset_index()

shopping_list
```

```python
complete_dates = shopping_list.groupby('quote_date').filter(lambda x: set(groceries_list.keys()).issubset(set(x['item_id'])))

earliest_complete_date = complete_dates['quote_date'].min()

last_complete_date = complete_dates['quote_date'].max()
```

```python
shopping_list = shopping_list[(shopping_list['quote_date'] >= earliest_complete_date) & (shopping_list['quote_date'] <= last_complete_date)]
```

```python
shopping_list['item_name'] = shopping_list['item_id'].map(groceries_list)

shopping_list
```

```python
shopping_list.to_csv('shopping_list.csv', index=False)
```
