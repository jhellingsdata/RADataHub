```python
import os
from typing import Iterable
import altair as alt
import numpy as np
import pandas as pd
import requests
import eco_style
alt.theme.enable("light")
```

```python
prices_df = pd.read_csv("grocery_prices.csv")
prices_df["quote_date"] = pd.to_datetime(prices_df["quote_date"])
```

```python
prices_df = prices_df[~prices_df["item_name"].isin(["Fish", "Chicken"])]
```

```python
hover = alt.selection_point(
    fields=["item_name"],
    on="pointerover",
    value=[{"item_name": "None"}]
)

chart = (
    alt.Chart(prices_df)
    .mark_area(interpolate="monotone")
    .encode(
        x=alt.X(
            "yearmonth(quote_date):T",
            axis=alt.Axis(format="%Y", title=None)
        ),
        y=alt.Y(
            "sum(price):Q",
            title=None,
            axis=alt.Axis(
                labelExpr='"£" + datum.value',
                titleAngle=0,
                titleAlign="left",
                titleBaseline="bottom",
                titleX=0,
                titleY=-10,
            )
        ),
        color=alt.Color(
            "item_name:N",
            legend=alt.Legend(title=None)
        ),
        opacity=alt.condition(hover, alt.value(1.0), alt.value(0.7)),
        tooltip=[
            alt.Tooltip("yearmonth(quote_date):T", title="Month", format="%b %Y"),
            alt.Tooltip("item_name:N", title="Item"),
            alt.Tooltip("sum(price):Q", title="Price (£)", format=",.2f"),
        ]
    )
    .add_params(hover)
    .properties(
        width=600,
        height=380,
        title=alt.TitleParams(
            text="The Rising Cost of a Basket of Essentials",
            subtitle=[
                "Average UK supermarket price of 6 staple grocery items",
                "Source: Food AutoCPI (Davies, McEvoy and Hellings)",
                ""
            ],
            anchor="start",
            fontSize=16,
            subtitleFontSize=12,
            subtitleColor="#676A86",
        )
    )
)


chart
```

```python
chart.save('grocery_prices.json')
chart.save('grocery_prices.png', scale_factor=3.0)
```
