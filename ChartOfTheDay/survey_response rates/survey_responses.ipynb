```python
survey = pd.read_excel('data.xlsx')
survey.dtypes
survey['year'] = pd.to_datetime(survey['year'], format='%Y')
```

```python

# Filter for the regions of interest

df_UK = survey[survey['country'] == 'UK']
df_US = survey[survey['country'] == 'US']
df_rest = survey[~survey['country'].isin(['UK', 'US'])]

# Get the last available value for each region
last_values_US = df_US.sort_values('year').groupby('country').last().reset_index()
last_values_UK = df_UK.sort_values('year').groupby('country').last().reset_index()
```

```python
# Interpolate missing values in df_rest
df_rest['value'] = df_rest['value'].interpolate()

# Remove values for year 2023
df_rest = df_rest[df_rest['year'] != '2023']

line_chart = alt.Chart(df_rest).mark_line(strokeWidth=2, color="#C8BCB4").encode(
    x=alt.X('year:T'),
    y=alt.Y('value', scale=alt.Scale(domain=[0, 100]), axis=alt.Axis(title=None, values=[0, 20, 40, 60, 80, 100], labelExpr="datum.value+'%'")),
    detail='country:N', 
    tooltip=['year:T', 'country:N', 'value:Q']
).properties(
    width=500,
    height=300
)

# Create the line chart for the UK trend
uk_line = alt.Chart(df_UK).mark_line(strokeWidth=2, color="#CE3140").encode(
    x=alt.X('year:T'),
    y=alt.Y('value:Q'),
    tooltip=['year:T', 'country:N', 'value:Q']
)

# Create the line chart for the US trend
us_line = alt.Chart(df_US).mark_line(strokeWidth=2, color="#208FCE").encode(
    x=alt.X('year:T'),
    y=alt.Y('value:Q'),
    tooltip=['year:T', 'country:N', 'value:Q']
)

# Add text labels for the UK trend at the last value point
text2 = alt.Chart(last_values_UK).mark_text(align='left', dx=5, dy=11).encode(
    x='year:T',
    y='value:Q',
    text='country:O',
    color=alt.value("#E6224B"),
    tooltip=['year:T', 'country:N', 'value:Q']
)

# Add text labels for the US trend at the last value point
text1 = alt.Chart(last_values_US).mark_text(align='left', dx=-30, dy=4).encode(
    x='year:T',
    y='value:Q',
    text='country:O',
    color=alt.value("#208FCE"),
    tooltip=['year:T', 'country:N', 'value:Q']
)

final_chart = line_chart + uk_line + us_line + text1 + text2

final_chart.save("survey_response.json")
final_chart.save("survey_response.png", scale_factor=3.0)
final_chart.save("survey_response.svg")

final_chart
```
