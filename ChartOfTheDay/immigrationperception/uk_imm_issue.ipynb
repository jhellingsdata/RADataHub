```python
import pandas as pd
import numpy as np
import altair as alt

# Using ecostyles package (installed in environment)
from ecostyles import EcoStyles
styles = EcoStyles()
styles.register_and_enable_theme(theme_name='article')
```

```python
uk_imm_issue = pd.read_csv('data-GnXxN.csv')
```

```python
uk_imm_issue.head()
```

```python
scatter = (
    alt.Chart(uk_imm_issue)
    .mark_circle(size=120, opacity=0.8, color='#4c78a8')
    .encode(
        x=alt.X('Share of population born abroad (%)', title='Share of population born abroad (%)'),
        y=alt.Y('Immigration is most important national problem (%)', title='Immigration is most important national problem (%)'),
        tooltip=[(alt.Tooltip("COUNTRYNEW", title="Country")), 'Share of population born abroad (%)', 'Immigration is most important national problem (%)']
    ).properties(
        width=800,
        height=350,
        title=alt.TitleParams(
            text='Immigration issue salience compared with foreign born share of population',
            subtitle='Poll data from Gallu (2025). Migration data from UN DESA (2024) ',
            subtitleFontSize=12,
            anchor='start'
        )
    )
)

highlight_countries = ['United Kingdom', 'United States', 'Ireland', 'Netherlands', 'Italy']

labels = (
    alt.Chart(uk_imm_issue)
    .transform_filter(alt.FieldOneOfPredicate(field='COUNTRYNEW', oneOf=highlight_countries))
    .mark_text(align='left', fontSize=12, clip=False, dx=8)
    .encode(
        x='Share of population born abroad (%)',
        y='Immigration is most important national problem (%)',
        text='COUNTRYNEW'
    )
)

alt.layer(scatter, labels).properties(width=400, height=400)
```
