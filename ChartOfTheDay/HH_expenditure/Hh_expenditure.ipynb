```python
consumption_pp = pd.read_excel('v2regionalhouseholdexpenditure2009to2022.xlsx', sheet_name="Sheet3")
consumption_pp
```

```python
df = pd.melt(consumption_pp, 
                  id_vars=['ITL level', 'ITL code', 'Region name', 'COICOP code', 'Commodity'],
                  var_name='Year', 
                  value_name='Value')

df

#keep only Wales, Scotland, Northern Ireland, England and UK
df = df[df['Region name'].isin(['Wales', 'Scotland', 'Northern Ireland', 'England', 'United Kingdom'])]
df
```

```python
#rename Region name United Kingdom to UK
df['Region name'] = df['Region name'].replace({'United Kingdom': 'UK'})

df['Value'] = df['Value'].astype(float)
df.dtypes
```

```python

# Filter for the regions of interest
df_land = df[df['Region name'].isin(['England', 'London', 'Wales', 'Scotland', 'Northern Ireland'])]
df_UK = df[df['Region name'] == 'UK']

# Get the last available value for each region
last_values = df_land.sort_values('Year').groupby('Region name').last().reset_index()
last_values_UK = df_UK.sort_values('Year').groupby('Region name').last().reset_index()
```

```python
line_chart = alt.Chart(df_land).mark_line(strokeWidth=1).encode(
    x=alt.X('Year:T'),
    y=alt.Y('Value', axis=alt.Axis(title=None, labelExpr="datum.value+'£'"), scale=alt.Scale(domain=[400, 1600] )),
    color=alt.Color('Region name:O', scale=alt.Scale(range=["#36B7B4","#EB5C2E","#F4C245","#0063AF","#00A767"]), legend=None),
    tooltip=['Year', 'Region name', 'Value']
).transform_filter(
    alt.datum['Region name'] != 'UK'  
).properties(
    title={
        "text": ["UK: Regional trends in clothing expenditure"],
        "subtitle": ["National household clothing and footwear expenditure in £ per person, 2009 - 2022", "Source: ONS 2025", ""],
        "subtitleColor": "#676A86",
        "anchor": "start",
        "fontSize": 22 
    }
)

# Add text labels for each region at the last value point
text = alt.Chart(last_values).mark_text(align='left', dx=5, dy=-2).encode(
    x='Year:T',
    y='Value:Q',
    text='Region name:O',
    color='Region name:O'
).transform_filter(
    alt.datum['Region name'] != 'UK'
)

# Add text labels for the UK trend at the last value point
text2 = alt.Chart(last_values_UK).mark_text(align='left', dx=5, dy=11).encode(
    x='Year:T',
    y='Value:Q',
    text='Region name:O',
    color=alt.value("#E6224B"),  
)

# Create the line chart for the UK trend
uk_line = alt.Chart(df_UK).mark_line(strokeDash=[1, 1], strokeWidth=2).encode(
    x=alt.X('Year:T'),
    y=alt.Y('Value:Q'),
    color=alt.value("#E6224B"),  
    tooltip=['Year', 'Region name', 'Value']
).transform_filter(
    alt.datum['Region name'] == 'UK'  
)


# Combine all parts into the final chart
final_chart = line_chart + uk_line + text + text2

# Display the final chart
```

```python
final_chart.save("clothing.json")
final_chart.save("clothing.png", scale_factor=3.0)
final_chart.save("clothing.svg")
final_chart
```
