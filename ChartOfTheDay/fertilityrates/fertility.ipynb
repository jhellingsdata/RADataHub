```python
#read csv fertility
df = pd.read_csv('fertility.csv')
df
```

```python
# Ensure Year is datetime for plotting
df['date'] = pd.to_datetime(df['Year'], format='%Y')
df['value'] = df['World']

df['highlight'] = False
df.iloc[0, df.columns.get_loc('highlight')] = True

# Create chart
chart = alt.Chart(df).mark_line(
    color="#00A767",
    point=alt.OverlayMarkDef(
        filled=False,
        opacity=alt.expr("if(datum.highlight, 1, 0)"),
        stroke="#00A767",
        fill="#00A767"
    )
).encode(
    x=alt.X('yearmonth(date):T', title=''),
    y=alt.Y('value:Q',
            scale=alt.Scale(domain=[0, 6]), title='' ),
).properties(
    title={
        "text": "Global Fertility Rates",
        "frame": "group",
        "anchor": "start",
        "subtitle": ["Total births per woman, 1960 - 2022", "Source: World Bank", ""],
        "subtitleColor": "#676A86",
        "fontSize": 22
    }
)
# Callout label for last point
callout_text = chart.transform_filter(alt.datum.highlight).mark_text(
    align='left',
    baseline='middle',
    dx=5,
    color='#00A767',
    fontSize=12,
    fontWeight='bold',
    text=alt.expr("format(datum.value, '.1f') + '%'")
)

# Final chart with annotation
chart0 = chart + callout_text

chart0.save("fertility.json")
chart0.save("fertility.png", scale_factor=3.0)
chart0.save("fertility.svg")
chart0

chart0
```

```python
# Melt to long format
df_long = df.melt(id_vars='Year', var_name='Country', value_name='value')

# Convert Year to datetime
df_long['date'] = pd.to_datetime(df_long['Year'], format='%Y')

# Highlight first year (latest data point per country, assuming descending order)
df_long = df_long.sort_values(['Country', 'Year'])
df_long['highlight'] = (df_long['Year'] == 2022)

# Color mapping
color_scale = alt.Scale(
    domain=["World", "United Kingdom", "United States", "Kenya", "Japan"],
    range=["#36B7B4","#E6224B","#F4C245","#0063AF","#00A767"]
)

# Main line chart
chart = alt.Chart(df_long).mark_line(
    point=alt.OverlayMarkDef(
        filled=False,
        opacity=alt.expr("if(datum.highlight, 1, 0)"),
    )
).encode(
    x=alt.X('year(date):T', title=''),
    y=alt.Y('value:Q', scale=alt.Scale(domain=[0, 10]), title=''),
    color=alt.Color('Country:N', scale=color_scale, legend=None),
    stroke=alt.Stroke('Country:N', scale=color_scale, legend=None),
)

# Callout labels with country names
callout = alt.Chart(df_long).transform_filter(
    alt.datum.highlight
).mark_text(
    align='left',
    baseline='middle',
    dx=5,
    dy=alt.ExprRef(expr="""
            {   'United States': -6,
            }[datum.Country]
        """),
    fontSize=12,
    fontWeight='bold'
).encode(
    x='year(date):T',
    y='value:Q',
    text='Country:N',
    color=alt.Color('Country:N', scale=color_scale)
)

# Combine chart and labels
final_chart = (chart + callout).properties(
    title={
        "text": "Global Fertility Rates",
        "frame": "group",
        "anchor": "start",
        "subtitle": ["Total births per woman, 1960 - 2022", "Source: World Bank", ""],
        "subtitleColor": "#676A86",
        "fontSize": 22
    }
)

# Save outputs
final_chart.save("fertility_country_labels.json")
final_chart.save("fertility_country_labels.png", scale_factor=3.0)
final_chart.save("fertility_country_labels.svg")

final_chart
```
