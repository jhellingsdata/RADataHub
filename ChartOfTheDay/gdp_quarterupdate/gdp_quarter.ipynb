```python
import altair as alt
import pandas as pd

# Data
data = pd.DataFrame({
    'Subsector': ['Subsectors affected by 25% tariffs']*3 + 
                 ['Subsectors affected by 10% tariffs']*3 + 
                 ['Subsectors not subject to tariffs']*3,
    'Demand Type': ['Domestic demand', 'EU demand', 'Non-EU demand']*3,
    'Percentage': [58, 13, 30, 85, 7, 8, 72, 17, 11]
})

# Add % label
data['Percentage Label'] = data['Percentage'].astype(str) + '%'

# Custom colors to match the image
custom_colors = {
    'Domestic demand': '#bce3e1',
    'EU demand': '#86c7a8',
    'Non-EU demand': '#7db4e9'
}

# Define order of bars (x-axis)
bar_order = [
    'Subsectors affected by 25% tariffs',
    'Subsectors affected by 10% tariffs',
    'Subsectors not subject to tariffs'
]

# Bar chart
chart = alt.Chart(data).mark_bar().encode(
    x=alt.X('Subsector:N', sort=bar_order, title=''),
    y=alt.Y('Percentage:Q', stack='normalize', title=''),
    color=alt.Color('Demand Type:N',
                    scale=alt.Scale(domain=list(custom_colors.keys()), range=list(custom_colors.values())),
                    legend=alt.Legend(orient='top', direction='horizontal', title=None)),
    order=alt.Order('Demand Type', sort='ascending'),
    tooltip=['Subsector', 'Demand Type', 'Percentage']
).properties(
    title={
        "text": "UK Production Destinations ",
        "frame": "group",
        "anchor": "start",
        "subtitle": ["Categorization by the magnitude of US tariffs", "Source: ONS 2025", ""],
        "subtitleColor": "#676A86",
        "fontSize": 22
    }, 
    width=500,
    height=300
)

# % labels for Non-EU demand
text = alt.Chart(data[data['Demand Type'] == 'Non-EU demand']).mark_text(
    align='center',
    baseline='middle',
    dy=11,
    fontSize=13
).encode(
    x=alt.X('Subsector:N', sort=bar_order),
    y=alt.Y('Percentage:Q', stack='normalize'),
    text='Percentage Label:N'
)

# Combine
final = chart + text
final
```

```python
final.save("prouction.json")
final.save("prouction.png", scale_factor=3.0)
final.save("prouction.svg")
```

```python
data = pd.DataFrame({
    'Month': [
        'Feb 2024',  'Mar 2024', 'Apr 2024', 'May 2024', 'Jun 2024', 'Jul 2024', 
        'Aug 2024', 'Sep 2024', 'Oct 2024', 'Nov 2024', 'Dec 2024', 'Jan 2025', 'Feb 2025'
    ],
    'GDP': [0.0043, 0.0091, 0.0087, 0.0088, 0.0044, 0.0023, -0.0002, 0.0000, -0.0001, -0.0004, 0.0006, 0.0026, 0.0061],
    'Construction': [-0.0003, 0.0000, -0.0004, 0.0000, 0.0002, 0.0007, 0.0005, 0.0002, 0.0000, 0.0001, 0.0002, 0.0001, 0.0000],
    'Services': [0.0046, 0.0088, 0.0087, 0.0089, 0.0049, 0.0027, 0.0000, 0.0002, -0.0001, 0.0001, 0.0009, 0.0029, 0.0052],
    'Production': [0.0000, 0.0003, 0.0004, -0.0001, -0.0006, -0.0012, -0.0007, -0.0004, -0.0001, -0.0007, -0.0006, -0.0004, 0.0009]
})
```

```python

# Melt for bar chart
bars_df = data.melt(id_vars='Month', value_vars=['Construction', 'Services', 'Production'],
                    var_name='Sector', value_name='Value')

# Prepare GDP for line with Sector label for color/legend
gdp_df = data.copy()
gdp_df['Sector'] = 'GDP'

gdp_label_df = gdp_df[gdp_df['Month'] == 'Feb 2025'].copy()
gdp_label_df['Label'] = 'GDP: 0.6%'

# Bar chart
bar = alt.Chart(bars_df).mark_bar().encode(
    x=alt.X('Month:N', sort=list(data['Month']), title='', 
            axis=alt.Axis(values=data['Month'][::3])),  # Show every 3rd value
    y=alt.Y('Value:Q', title='', axis=alt.Axis(format=".1%")),
    color=alt.Color('Sector:N',
                    legend=alt.Legend(orient='top', direction='horizontal', title=None),
                    scale=alt.Scale(domain=['Production', 'Construction', 'Services', 'GDP'],
                                    range=["#bce3e1", "#7db4e9", "#86c7a8", "#1f78b4"])),
    tooltip=['Month', 'Sector', alt.Tooltip('Value:Q', format=".1%")]
)

# Line chart with points for GDP
line = alt.Chart(gdp_df).mark_line(point=True).encode(
    x=alt.X('Month:N', sort=list(data['Month'])),
    y=alt.Y('GDP:Q'),
    color=alt.Color('Sector:N',
                    scale=alt.Scale(domain=['GDP'], range=['#1f78b4']),
                    legend=alt.Legend(orient='top', direction='horizontal', title=None)),
    tooltip=['Month', alt.Tooltip('GDP:Q', format=".1%")]
)

# Horizontal line at y = 0
hline = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='#676A86').encode(
    y='y:Q'
)
# Text label for GDP
gdp_label = alt.Chart(gdp_label_df).mark_text(
    align='left',
    baseline='middle',
    dx=-5,
    dy=-10,
    fontSize=12,
    color='#1f78b4'
).encode(
    x=alt.X('Month:N', sort=list(data['Month'])),
    y=alt.Y('GDP:Q'),
    text='Label:N'
)

# Final chart
gdp_chart = (hline + bar + line + gdp_label).properties(
    title={
        "text": "UK: GDP Contribution by Sector",
        "frame": "group",
        "anchor": "start",
        "subtitle": ["Three months averages, Feb 2024 - Feb 2025", "Source: ONS 2025", ""],
        "subtitleColor": "#676A86",
        "fontSize": 22
    },
    width=600,
    height=300
)

gdp_chart
```

```python
gdp_chart.save("gdp.json")
gdp_chart.save("gdp.png", scale_factor=3.0)
gdp_chart.save("gdp.svg")
```
