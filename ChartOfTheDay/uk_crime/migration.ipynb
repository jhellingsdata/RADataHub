```python
df = pd.read_excel("datadownload.xlsx", sheet_name="data")
df_latest = df[df['Year'].isin(['Dec 2023', 'Dec 2024'])].copy()
```

```python
df = df.set_index('Year')
baseline = df.iloc[0]
indexed_df = df.apply(lambda col: (col / baseline[col.name]) * 100)
indexed_df = indexed_df.reset_index()
```

```python
long_df = indexed_df.melt(id_vars='Year', var_name='type', value_name='Index')
latest_values = long_df[long_df['Year'] == "Dec 2024"]
latest_values
```

```python
year_order = indexed_df['Year'].tolist()
```

```python

long_df = long_df.sort_values(by=['type', 'Year']).reset_index(drop=True)
long_df['Index'] = (
    long_df
    .groupby('type')['Index']
    .apply(lambda group: group.interpolate(method='linear'))
)
```

```python
chart = (
    alt.Chart(long_df)
    .mark_line(point=False)
    .encode(
        x=alt.X('Year', sort=year_order),
        y=alt.Y('Index:Q', title='Index (2017 = 100)'),
        color= alt.Color('type:N', legend=None),
        tooltip=['Year', 'type', 'Index']
    ))

text = alt.Chart(latest_values).mark_text(align='left', dx=5, 
                                        dy=alt.expr('{"Knife and sharp instruments":5.2,}[datum.type]')).encode(
    x=alt.X('Year', sort=year_order),
    y='Index:Q',
    text='type:O',
    color='type:O'
)


final_chart = chart + text

final_chart = final_chart.properties(
    title={
        "text": ["Crime in England and Wales"],
        "subtitle": ["Incidents and offences by crime type, 2017 - 2024 (Mar 2017 = 100)", "Source: CSEW and Home Office 2025", ""],
        "subtitleColor": "#676A86",
        "anchor": "start",
        "fontSize": 22 
    }
)

final_chart
```

```python
df_latest = df_latest.set_index('Year')
baseline = df_latest.iloc[0]
indexed_df = df_latest.apply(lambda col: (col / baseline[col.name]) * 100)
indexed_df = indexed_df.reset_index()
long_df_latest = indexed_df.melt(id_vars='Year', var_name='type', value_name='Index')
#long_df_latest
```

```python
chart = alt.Chart(long_df_latest).mark_bar().encode(
    y=alt.Y('Index:Q', title=None),
    x=alt.X('type:N',title=None),
    xOffset='Year:N', 
    color=alt.Color('Year:N', title=None),
)

hline = alt.Chart(pd.DataFrame({'y': [100]})).mark_rule(
    color='black', strokeDash=[5, 5]).encode(
    y='y:Q' 
)

final_chart_bar = chart + hline

final_chart_bar = final_chart_bar.properties(
    width=600,
    height=400,
    title={
        "text": ["Crime in England and Wales"],
        "subtitle": ["Incidents by crime type (Dec 2023= 100)", "Source: Crime Survey for England and Wales", ""],
        "subtitleColor": "#676A86",
        "anchor": "start",
        "fontSize": 20}
    ).configure_legend(
    orient='top',
    direction='horizontal',
    symbolType='square'
)
final_chart_bar
```
