```python
# convert the vechicles data to long form
vehicles_long = vehicles.melt(id_vars='Month', var_name='Fuel', value_name='Percent')
vehicles_long
```

```python
vehicles['total'] = vehicles.sum(axis=1)
vehicles
```

```python
# make a stacked area chart with altair
chart = alt.Chart(vehicles_long,
                  title=alt.Title('Second-Hand Vehicles Sales by Fuel Type in the UK',
                                  subtitle='June 2021 - June 2024. Data Source: ONS.')
                  ).mark_area().encode(
    x=alt.X('Month:T', title='Month', axis=alt.Axis(format='%b %Y')),
    y=alt.Y('Percent:Q', title=None).stack('normalize'),
    # color by fuel type and order by the nominal range
    color=alt.Color('Fuel:N', title='Fuel Type', scale=alt.Scale(
        domain=['Hybrid', 'Diesel', 'Petrol'],
        range=[pallete['nominal_1'], pallete['nominal_2'], pallete['nominal_3']]
    )),
    tooltip=[
        alt.Tooltip('Month:T', title='Month', format='%B %Y'),
        alt.Tooltip('Percent:Q', title='Percent', format='.0%'),
        alt.Tooltip('Fuel:N', title='Fuel Type')
    ]
).properties(
    width=800,
    height=400
)
# make the leged horizontal and at the top
chart = chart.configure_legend(
    orient='top',
    titleFontSize=12,
    labelFontSize=11
)
chart
```

```python
chart.save('ons_vehicles.json')
chart.save('ons_vehicles.png', scale_factor=2)
```
