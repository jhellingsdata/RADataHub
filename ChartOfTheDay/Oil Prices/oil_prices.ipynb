```python
import os
from typing import Iterable
import altair as alt
import numpy as np
import pandas as pd
import requests
import eco_style
alt.theme.enable("light")
```

```python
df = pd.read_csv('oil_prices.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
end_date = df['Date'].max()
start_date = end_date - pd.DateOffset(months=6)
df_filtered = df[df['Date'] >= start_date].copy()
```

```python

conflict_date = pd.to_datetime('2026-02-28')

base = alt.Chart(df_filtered).encode(
    x=alt.X(
        'Date:T', 
        title=None, 
        axis=alt.Axis(
            format='%b %Y', 
            labelAngle=0, 
            tickCount="month"  
        )
    ),
    y=alt.Y(
        'Value:Q',
        title=None,
        scale=alt.Scale(domain=[59, 100]), 
        axis=alt.Axis(format='$,.0f')
    )
)

line = base.mark_line(
    strokeWidth=2,
    color='#179fdb',
).properties(
    title=alt.TitleParams(
        text='Brent Crude Oil Price Surge (USD/Bbl)',
        subtitle=[f"Spike corresponds with Operation Epic Fury | Source: Macrotrends & Trading Economics"],
        fontSize=15,
        subtitleFontSize=12,
        subtitleColor="#676A86",
        color="#676A86",
        anchor='start'
    ),
    width=600,
    height=400,
)

rule_df = pd.DataFrame({'Date': [conflict_date]})
rule = (
    alt.Chart(rule_df)
    .mark_rule(color='#e6224b', strokeDash=[5, 5], strokeWidth=1)
    .encode(x='Date:T')
)

annotation_df = pd.DataFrame({'Date': [conflict_date], 'Label': [None]})
text = (
    alt.Chart(annotation_df)
    .mark_text(
        align='right',
        dx=-3,
        dy=0,
        fontWeight='bold',
        fontSize=10,
        color='#e6224b'
    )
    .encode(
        x='Date:T',
        y=alt.value(50),
        text='Label:N'
    )
)

chart = (
    (line + rule )
    .configure_view(strokeWidth=0)
    .configure_axis(
        grid=True,          
        gridColor='#eeeeee', 
        domain=True,       
        domainColor='#676A86', 
        tickColor='#676A86'
    )
    .configure_axisX(
        grid=False          
    )
    .configure_title(
        fontSize=16, 
        subtitleFontSize=12, 
        subtitleColor='gray'
    )
)

chart.display()
```

```python
chart.save('oil_prices.json')
chart.save('oil_prices.png', scale_factor=3.0)
```
