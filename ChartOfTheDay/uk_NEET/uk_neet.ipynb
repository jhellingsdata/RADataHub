```python
data = pd.read_excel('notineducationemploymentortrainingjm25.xlsx', sheet_name="data")
data
```

```python
# Convert the 'date' column to a datetime type using the last month in the range as the date
def parse_quarter_date(s):
    # Example: "Jan-Mar 2025" -> "Mar 2025"
    try:
        months, year = s.split()
        last_month = months.split('-')[-1]
        return pd.to_datetime(f"{last_month} {year}")
    except Exception:
        return pd.NaT

data['date_dt'] = data['date'].apply(parse_quarter_date)
```

```python
data = data.melt(id_vars='date_dt', value_vars=['Men',  'Women'], #Total',
                        var_name='group', value_name='percentage')

#divide percentage by 1000
data['percentage'] = data['percentage'] / 1000
```

```python
#create new df last values based on Jan-Mar 2025 in date
last_values = data[data['date_dt'] == '2025-03-01'].copy()
last_values 
```

```python
line_chart = alt.Chart(data).mark_line(strokeWidth=1).encode(
    x=alt.X('date_dt'),
    y=alt.Y('percentage', axis=alt.Axis(title=None, labelExpr="datum.value+'K'")),#, scale=alt.Scale(domain=[0, 15] )
    color=alt.Color('group:O', scale=alt.Scale(range=["#36B7B4","#EB5C2E","#00A767"]), legend=None)
).transform_filter(
    alt.datum['group'] != 'Total'  
).properties(
    title={
        "text": ["UK: Young people not in education, employment or training "],
        "subtitle": ["Men and women aged 16 to 24 years, seasonally adjusted, Jan-Mar 2019 to Jan-Mar 2025", "Source: ONS 2025", ""],
        "subtitleColor": "#676A86",
        "anchor": "start",
        "fontSize": 20
    }
)
line_chart

# Add text labels for each region at the last value point
text = alt.Chart(last_values).mark_text(align='left', dx=5).encode(
    x='date_dt',
    y='percentage',
    text='group',
    color='group'
).transform_filter(
    alt.datum['group'] != 'Total'
)

final_chart = line_chart + text
final_chart
```

```python
final_chart.save("neet.json")
final_chart.save("neet.png", scale_factor=3.0)
final_chart.save("neet.svg")
```
