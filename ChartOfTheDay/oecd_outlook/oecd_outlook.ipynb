```python
data = pd.read_excel('chartdata.xlsx', sheet_name="Sheet1")
data
```

```python
#rremove the column difference
data = data.drop(columns=["Difference"])
df_melted = data.melt(id_vars='entity', var_name='Projection Type', value_name='Value')
```

```python
#remove rows with entity = Korea, indonesia, and China
df_melted = df_melted[~df_melted['entity'].isin(['Korea', 'Indonesia', 'Russia', 'South Africa', 'Spain', 'Australia'])]
```

```python
entity_order
```

```python
# Step 1: Determine sort order by mean or any metric of 'Value'
entity_order = df_melted.groupby('entity')['Value'].mean().sort_values(ascending=False).index.tolist()
```

```python
# Proper grouped horizontal bar chart
chart = alt.Chart(df_melted).mark_bar().encode(
    x=alt.X('Value:Q', title='2025 Projection (%)', axis=alt.Axis(title=None, labelExpr="datum.value+'%'")),
    yOffset='Projection Type:N', 
    y=alt.Y('entity:N', sort=entity_order, title=None),
    color=alt.Color('Projection Type:N', title=None),
    tooltip=['entity', 'Projection Type', 'Value']
)

width = 400

rect = alt.Chart(df_melted[df_melted['entity'] == 'World']).mark_rect(opacity=0.4).encode(
    alt.Y('entity:N', sort=entity_order,title=None),
    alt.YOffset('Projection Type:N'),
    x=alt.value(0),
    x2=alt.value(width)
)

final_chart = (chart + rect).properties(
    title={
        "text": ["Changes in 2025 GDP growth projections"],
        "subtitle": ["OECD forecasts for GDP growth, June 2025 vs December 2024 projection", "Source: OECD Economic Outlook", ""],
        "subtitleColor": "#676A86",
        "anchor": "start",
        "fontSize": 20
    }
).configure_legend(
    orient='top',
    direction='horizontal',
    symbolType='square'
)

final_chart
```

```python
final_chart.save("gdp.json")
final_chart.save("gdp.png", scale_factor=3.0)
final_chart.save("gdp.svg")
```
