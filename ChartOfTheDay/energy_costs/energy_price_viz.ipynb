```python
url = 'https://raw.githubusercontent.com/jhellingsdata/RADataHub/main/ChartOfTheDay/energy_costs/costs_long.csv'
```

```python
data = pd.read_csv(url)
```

```python
# change date from string in format M/DD/YY to datetime
data['Date'] = pd.to_datetime(data['Date'], format='%m/%d/%y')
```

```python
data
```

```python
# eliminate all rows where the value is not a number
data = data[data['Value'].notnull()]
#eliminate all rows where 'Cost' is 'Total'
data = data[data['Cost'] != 'Total']
```

```python
# restrict to only data since 2020
data = data[data['Date'] >= '2020-01-01']
```

```python
import altair as alt
# Enable dark theme
alt.themes.enable('report')

# Define the base chart with the new tooltip and custom color scheme
chart_interactive = alt.Chart(data,
                 title=alt.Title("Price Cap Breakdown",
                                 subtitle=["Costs for dual-fuel, prepayment-meter households, 2020-2024. Source: Ofgem.",""],
                                 anchor = 'start')
                 ).mark_area(interpolate = 'step-after').encode(
    x=alt.X('Date:T',
            title='Time Period',
            axis=alt.Axis(format='%b %Y', labelAngle=0)),
    y=alt.Y('Value:Q', 
            title='Cost (GBP)',
            axis=alt.Axis(format='~s',
                          labelExpr="'£' + format(datum.value, ',')")),
    color=alt.Color('Cost:N', 
                    scale=alt.Scale(
                        range=[
    "#36B7B4",
    "#E6224B",
    "#243B5A",
    "#0063AF",
    "#00A767",
    "#179FDB",
    "#C9C9C9",
    "#F4C245",
    "#EB5C2E"
]
                    ),
                    legend=alt.Legend(
                        orient='right',
                        title=None,  
                        symbolSize=200, 
                        symbolStrokeWidth=3
                    )),
    tooltip=[
        alt.Tooltip('Cost:N', title='Category'),
        alt.Tooltip('Value:Q', title='Cost (GPB)', format=',')
    ]
)

# set the interactive opacity of the chart
highlight = alt.selection(type='single', on='mouseover', empty='none', fields=['Cost'])
opacity = alt.condition(highlight, alt.value(1), alt.value(0.7))

# Adjust the properties of the chart
chart_interactive = chart_interactive.add_selection(
    highlight
).encode(
    opacity=opacity
).properties(
    width=800,
    height=400
)

chart_interactive
```
