##### Table 1b: UK government expenditure on R&D in percentage share

Sheet 8, skip 11 rows

```python
df = pd.read_excel('goverd2022.xlsx', sheet_name='8', skiprows=11)
```

```python
# Remove `TOTAL` row
df = df[df['Socio-economic objective'] != 'TOTAL'].reset_index(drop=True)

# Replace '[x]' with 0
df = df.replace('[x]', 0)

df
```

```python
# Melt the dataframe
df_long = df.melt(id_vars=['Socio-economic objective'], var_name='date', value_name='value')

df_long['date'] = df_long['date'].astype(str) + '-01-01'
```

```python
df
```

```python
alt.Chart(df_long).mark_area().encode(
    alt.X('date:T'),
    alt.Y('value:Q', stack='normalize'),
    alt.Color('Socio-economic objective:N'),
)
```
