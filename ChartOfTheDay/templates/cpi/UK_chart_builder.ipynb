```python
import altair as alt
import pandas as pd
import numpy as np

# Sample data for demonstration purposes
np.random.seed(0)
dates = pd.date_range(start="2015-01-01", end="2023-01-01", freq='M')
values = np.random.uniform(0, 0.05, len(dates))  # Random values for demonstration

# Creating a sample DataFrame
df = pd.DataFrame({'date': dates, 'value': values})

# Dark theme flag
dark = True

# Colors based on the theme
background = '#122b39' if dark else '#b4c8d8'
detail = '#b4c8d8' if dark else '#122b39'
line_colour = '#e6224b'

# Creating a custom theme function
def custom_theme():
    # Configurations for the axes
    axis = alt.Axis(
        labelColor=detail,
        tickColor=detail,
        domainColor=detail,
        domainOpacity=0.5,
        grid=False,
        labelAngle=0,
        tickCount=10,
        tickOpacity=0.5,
        title=None
    )

    # Configurations for the y-axis with grid lines
    yAxis = alt.Axis(
        labelColor=detail,
        tickColor=detail,
        domainColor=detail,
        gridColor=detail,
        gridDash=[1, 5],
        gridOpacity=0.5,
        labelPadding=5,
        tickCount=8,
        tickOpacity=0.5,
        ticks=False,
        title='12-month percent change | Source: BLS via FRED',
        titleAlign='left',
        titleAngle=0,
        titleBaseline='bottom',
        titleColor=detail,
        titleFontSize=12,
        titleOpacity=0.9,
        titleX=0,
        titleY=-7,
        format='%'
    )

    # Returning the configurations as a dictionary
    return {
        "config": {
            "view": {
                "continuousWidth": 350,
                "continuousHeight": 280,
                "stroke": None
            },
            "background": background,
            "title": {
                "anchor": 'start',
                "dx": 24,
                "fontSize": 12,
                "subtitleFontSize": 12,
                "color": detail,
                "subtitleColor": detail
            },
            "axisX": axis,
            "axisY": yAxis,
            "line": {
                "color": line_colour
            },
            "font": "Circular Std"
        }
    }

# Register the custom theme under a name
alt.themes.register('custom_theme', custom_theme)

# Enable the theme
alt.themes.enable('custom_theme')

# Displaying the current theme to check if it's applied
alt.themes.active
```

```python
df = df.rename(columns={'Unnamed: 0': 'Division'})
```

```python
# Define lookup to clean `Division` names
lookup = {
    'Miscellaneous goods and services': 'Misc. goods and services',
    'CPI exc food, energy, alcohol and\ntobacco (core CPI)': 'Core CPI',
    'Housing and household services': 'Household services',
    'Furniture and household goods': 'Household goods',
    'Food and non-alcoholic beverages': 'Food and non-alc. drink',
}

# Replace any names in Division
df['Division'] = df['Division'].replace(lookup)
```

```python
df
```

```python
df['Sep 2024'] /= 100
df['Oct 2024'] /= 100

# Drop if NaN
df = df.dropna(subset=['Oct 2024'])
# Add a ranking column based on `Mar 2024` values
df['Rank'] = df['Oct 2024'].rank(ascending=False).astype(int)
df
```

```python
import altair as alt
```

```python
title=alt.TitleParams(
    "UK Inflation by Division",
    subtitle="CPI major divisons, 12-month percent change | Source: ONS",
    frame='group',
    dx=0
)

bars = alt.Chart(df, title=title).mark_bar().encode(
    alt.X('Oct 2024:Q').axis(format='%').scale(padding=20),
    alt.Y('Division:N').sort(alt.SortField('Rank', 'ascending')),
    color=alt.condition(alt.datum.Division == 'CPI All items', alt.value('#e6224b'), alt.value('#36b7b4'))
)

text_pos = bars.mark_text(
    align='left',
    baseline='middle',
    dx=5
).encode(
    alt.Text('Oct 2024:Q', format='.1%'),
    opacity=alt.condition(alt.datum['Oct 2024'] > 0, alt.value(1), alt.value(0)),
    color=alt.value('#b4c8d8')
)


text_neg = bars.mark_text(
    align='right',
    baseline='middle',
    dx=-4
).encode(
    alt.Text('Oct 2024:Q', format='.1%'),
    opacity=alt.condition(alt.datum['Oct 2024'] < 0, alt.value(1), alt.value(0)),
    color=alt.value('#b4c8d8')
)

chart = (bars + text_pos + text_neg)

chart.display()
```
