```python
import pandas as pd
```

```python
url = "https://happiness-report.s3.us-east-1.amazonaws.com/2025/Data+for+Figure+2.1+(2011%E2%80%932024).xlsx"
```

```python
df = pd.read_excel(url)
```

```python
whr_24 = df[df["Year"] == 2024]

whr_24
```

```python
whr_24.columns
```

```python
whr_24 = whr_24[['Country name','Rank','Year','Explained by: Log GDP per capita',
       'Explained by: Social support', 'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption',
       'Dystopia + residual']].copy()

whr_24
```

```python
# melt. index cols are 'Counrty name' and 'Rank'
whr_24 = pd.melt(whr_24, id_vars=['Country name','Rank'], var_name='Factor', value_name='Score')

whr_24
```

```python
# create a separate df with top 10 and bottom 10
whr_24_top = whr_24[whr_24['Rank'] <= 10].copy()

whr_24_bottom = whr_24[whr_24['Rank'] >= 138].copy()
```

```python
# add a column to indicate top or bottom
whr_24_top['Top/Bottom'] = 'Top 10'
whr_24_bottom['Top/Bottom'] = 'Bottom 10'
# concatenate top and bottom
whr_24_combined = pd.concat([whr_24_top, whr_24_bottom], ignore_index=True)


whr_24_combined
```

```python
whr_24_combined.to_csv('whr.csv', index=False)
```
