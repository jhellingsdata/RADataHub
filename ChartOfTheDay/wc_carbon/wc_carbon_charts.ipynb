```python
chart = alt.Chart(emissions_df).mark_bar().encode(
    x=alt.X('tournament:N', title=None, axis=alt.Axis(labelAngle=0)),
    y=alt.Y('emissions:Q', title='Emissions (tCO₂e)', axis=alt.Axis(grid=False)),
    color=alt.Color('category:N',
        sort=category_order,
        scale=alt.Scale(scheme='tableau10'),
        legend=alt.Legend(title='Category', orient='right')
    ),
    order=alt.Order('category:N', sort='ascending')
).properties(
    width=400,
    height=450
)

chart
```
