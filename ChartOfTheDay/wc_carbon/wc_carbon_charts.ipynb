```python
from ecostyles import EcoStyles
# Create styles instance
styles = EcoStyles()

import altair as alt
import pandas as pd
```

```python
# Register and enable a theme
styles.register_and_enable_theme(theme_name="article")
```

```python
# Create dataframe
wc_carbon_df = pd.DataFrame([
    {'tournament': 'Qatar 2022', 'category': 'Travel (team + spectator)', 'emissions': 1878106},
    {'tournament': 'Qatar 2022', 'category': 'Infrastructure', 'emissions': 893337},
    {'tournament': 'Qatar 2022', 'category': 'Accommodation', 'emissions': 728403},
    {'tournament': 'Qatar 2022', 'category': 'Other', 'emissions': 131188},
    {'tournament': '2026 (est.)', 'category': 'Travel (team + spectator)', 'emissions': 6835439},
    {'tournament': '2026 (est.)', 'category': 'Infrastructure', 'emissions': 238105},
    {'tournament': '2026 (est.)', 'category': 'Accommodation', 'emissions': 367200},
    {'tournament': '2026 (est.)', 'category': 'Other', 'emissions': 326800}
])

# Define category order -- travel at top so it's most prominent
category_order = ['Travel (team + spectator)', 'Accommodation', 'Infrastructure', 'Other']
```

```python
# Create World Cup emissions bar chart
wc_carbon_bar = alt.Chart(wc_carbon_df).mark_bar().encode(
    x=alt.X('tournament:N', title=None, axis=alt.Axis(labelAngle=0)),
    y=alt.Y(
        'emissions:Q', 
        title='Emissions (millions tCO₂e)', 
        axis=alt.Axis(
            grid=False,
            labelExpr="datum.value / 1000000 + 'm'")),
    color=alt.Color(
        'category:N',
        sort=category_order,
        legend=alt.Legend(title='Category', orient='right')
    ),
    order=alt.Order('category:N', sort='ascending')
).properties(
    title=alt.TitleParams(
        'Source: Greenly (2026), FIFA Greenhouse Gas Accounting Report (2022)',
        orient='bottom',
        fontStyle='italic',
        fontSize=10,
        color='#676A8680',
        fontWeight='normal',
        frame='group',
        offset=7
    )
)

# Define title layer
title = alt.Chart({'values': [{}]}).mark_text(
    text="The 2026 World Cup Will Emit More Than Double Qatar 2022",
    align='left',
    baseline='top',
    fontSize=14
).encode(
    x=alt.value(-30),
    y=alt.value(-35)
)

wc_carbon_bar = title + wc_carbon_bar
wc_carbon_bar
```

The 2026 FIFA World Cup is estimated to generate 7.8 million tonnes of CO₂, more than double Qatar 2022. Spectator and team travel accounts for 88% of the total. However, Qatar's central location meant shorter average travel distances, making direct comparisons difficult.

```python
# Save chart
styles.save(wc_carbon_bar, 'visualisation', 'wc_carbon_bar_chart', width=450, height=360)
```
