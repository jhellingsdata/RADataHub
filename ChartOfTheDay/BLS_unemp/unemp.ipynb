```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_csv("unemp_8_12_23.csv")
df = df.reset_index()
df['Value'] = df['Value'].astype(float)/100
df['Month'] = pd.to_datetime(df['Month'])
df = df[df['Month'] >= '2020-01-01']
```

```python
base = alt.Chart(df).encode(
    x=alt.X('Month:T',
                        axis=alt.Axis(
                title=None,
                labelColor="#b4c8d8",
                labelFont="Circular Std",
                gridColor="#b4c8d8",
                gridDash=[1, 5],
                gridOpacity=0,
                labelAlign="center",
                domainOpacity=0.5,
                labelOpacity=1,
                labelPadding=5,
                ticks=False,

            )

            
            ),
    y=alt.Y('Value:Q',
            axis=alt.Axis(
            title=None,
            gridColor="#b4c8d8",
            gridDash=[1, 5],
            labelFont="Circular Std",
            gridOpacity=0.5,            
            tickOpacity=0.5,
            format=".0%",
            domainOpacity=0.5,
            labelColor="#b4c8d8",
            labelOpacity=1,
            labelPadding=5,
            domainColor='#b4c8d8',
            tickColor='#b4c8d8'),
            
            
            ),
    color=alt.Color('Country:N', legend=None, scale=alt.Scale(reverse=False, range=[   
         "#36B7B4",
    "#00A767",
])),
)

bars = base.mark_line()

labels_1 = base.transform_filter("datum.index == 240").mark_text(
    align='left',
    baseline='middle',
    font='Circular Std',
    color='#b4c8d8',
    limit=48,
    ellipsis=".",
    dx = 5,
    dy=-5,
    fontSize=10,
    text=alt.expr("'November'"),
    opacity=1,
).encode(
    opacity=alt.value(1),
    color=alt.value('#b4c8d8'),
)

labels_2 = base.transform_filter("datum.index == 240").mark_text(
    align='left',
    baseline='middle',
    font='Circular Std',
    color='#b4c8d8',
    limit=48,
    ellipsis=".",
    dx = 5,
    dy=5,
    fontSize=10,
    text=alt.expr("'3.7%'"),
    opacity=1,
).encode(
    opacity=alt.value(1),
    color=alt.value('#b4c8d8'),
)


chart = bars + labels_1 + labels_2

chart = chart.properties(
    title={
        "text": ["US Unemploymnet Ticks Down to 3.7% in November"],
        "subtitle": ["Source: BLS through ECO API"],
                             "subtitleColor": "#b4c8d8",
        "fontSize": 12,
        "color": "#b4c8d8",
        "anchor": "start",
        "font": "Circular Std",
        "subtitleFont": "Circular Std",
        "dx": 28,
        "subtitleFontSize": 10,
    },
    width=300,
    height=200,
    background='#122b39',
).configure_view(
    strokeWidth=0
)

chart.save('us_unemp.json')
chart.save('us_unemp.png', scale_factor=3.0)

chart
```
