```python
import pandas as pd
import altair as alt
```

```python
gdp_for = pd.read_excel('Chapter_2_charts_and_tables_November_2025.xlsx', sheet_name='C2.10', skiprows=24)
```

```python
# Rename columns by combining current column names with first row values
gdp_for.columns = [f"{col}_{val}" for col, val in zip(gdp_for.columns, gdp_for.iloc[0])]
gdp_for = gdp_for.drop(0).reset_index(drop=True)

gdp_for.drop(columns=['Unnamed: 0_nan'], inplace=True)
```

```python
gdp_for.columns
```

```python
gdp_for.rename(columns={'Unnamed: 1_nan': 'Year', 
                        'March 2025 forecast_nan': 'March 2025 Forecast', 
                        'November 2025 forecast_nan': 'November 2025 Forecast',
                        'Percentile_10.0': 'Percentile_10',
                        'Unnamed: 5_20.0': 'Percentile_20',
                        'Unnamed: 6_30.0': 'Percentile_30',
                        'Unnamed: 7_40.0': 'Percentile_40',
                        'Unnamed: 8_50.0': 'Percentile_50',
                        'Unnamed: 9_60.0': 'Percentile_60',
                        'Unnamed: 10_70.0': 'Percentile_70',
                        'Unnamed: 11_80.0': 'Percentile_80',
                        'Unnamed: 12_90.0': 'Percentile_90'}, inplace=True)
```

```python
gdp_for['Year'] = gdp_for['Year'].astype(int)
```

```python
for i in range(len(gdp_for)):
    gdp_for['Percentile_20'].iloc[i] = gdp_for['Percentile_10'].iloc[i] + gdp_for['Percentile_20'].iloc[i]
    gdp_for['Percentile_30'].iloc[i] = gdp_for['Percentile_20'].iloc[i] + gdp_for['Percentile_30'].iloc[i]
    gdp_for['Percentile_40'].iloc[i] = gdp_for['Percentile_30'].iloc[i] + gdp_for['Percentile_40'].iloc[i]
    gdp_for['Percentile_50'].iloc[i] = gdp_for['Percentile_40'].iloc[i] + gdp_for['Percentile_50'].iloc[i]
    gdp_for['Percentile_60'].iloc[i] = gdp_for['Percentile_50'].iloc[i] + gdp_for['Percentile_60'].iloc[i]
    gdp_for['Percentile_70'].iloc[i] = gdp_for['Percentile_60'].iloc[i] + gdp_for['Percentile_70'].iloc[i]
    gdp_for['Percentile_80'].iloc[i] = gdp_for['Percentile_70'].iloc[i] + gdp_for['Percentile_80'].iloc[i]
    gdp_for['Percentile_90'].iloc[i] = gdp_for['Percentile_80'].iloc[i] + gdp_for['Percentile_90'].iloc[i]
```

```python
# Cleaner way using cumsum
percentile_cols = ['Percentile_10', 'Percentile_20', 'Percentile_30', 'Percentile_40', 
                   'Percentile_50', 'Percentile_60', 'Percentile_70', 'Percentile_80', 'Percentile_90']

gdp_for[percentile_cols] = gdp_for[percentile_cols].cumsum(axis=1)
```

```python
nov_forecast = alt.Chart(gdp_for).mark_line().encode(
    x=alt.X('Year:O', axis=alt.Axis(title=None, labelAngle=0 ,labelFontSize=14)),
    y=alt.Y('November 2025 Forecast:Q', axis=alt.Axis(title=None, labelFontSize=14)),
    color=alt.value('red')
).properties(
    title='OBR November 2025 GDP Growth Forecast',
    height=400,
    width=600
)

mar_forecast = alt.Chart(gdp_for).mark_line().encode(
    x='Year:O',
    y='March 2025 Forecast:Q',
    color=alt.value('green')
)

zero_line = alt.Chart(pd.DataFrame({'Year': gdp_for['Year'], 'y': [0]*len(gdp_for)})).mark_line(strokeDash=[5,5], color='black').encode(
    x='Year:O',
    y='y:Q'
)  

perc4060 = alt.Chart(gdp_for).mark_area(opacity=0.3, color='lightblue').encode(
    x='Year:O',
    y='Percentile_40:Q',
    y2='Percentile_60:Q'
)

perc3070 = alt.Chart(gdp_for).mark_area(opacity=0.3, color='blue').encode(
    x='Year:O',
    y='Percentile_30:Q',
    y2='Percentile_70:Q'
)

perc2080 = alt.Chart(gdp_for).mark_area(opacity=0.3, color='darkblue').encode(
    x='Year:O',
    y='Percentile_20:Q',
    y2='Percentile_80:Q'
)

perc1090 = alt.Chart(gdp_for).mark_area(opacity=0.3, color='grey').encode(
    x='Year:O',
    y='Percentile_10:Q',
    y2='Percentile_90:Q'
)

nov_forecast + mar_forecast + perc4060 + perc3070 + perc2080 + perc1090 + zero_line
```

```python
# Create the fan chart layers
# Fan areas (from darkest/narrowest to lightest/widest)
fan_90_10 = alt.Chart(gdp_for).mark_area(opacity=0.15, color='lightblue').encode(
    x=alt.X('Year:O', title='Year'),
    y=alt.Y('Percentile_10:Q', title='GDP Growth (%)'),
    y2='Percentile_90:Q'
)

fan_80_20 = alt.Chart(gdp_for).mark_area(opacity=0.2, color='lightblue').encode(
    x='Year:O',
    y='Percentile_20:Q',
    y2='Percentile_80:Q'
)

fan_70_30 = alt.Chart(gdp_for).mark_area(opacity=0.25, color='lightblue').encode(
    x='Year:O',
    y='Percentile_30:Q',
    y2='Percentile_70:Q'
)

fan_60_40 = alt.Chart(gdp_for).mark_area(opacity=0.3, color='lightblue').encode(
    x='Year:O',
    y='Percentile_40:Q',
    y2='Percentile_60:Q'
)

# Central line (50th percentile)
central_line = alt.Chart(gdp_for).mark_line(color='blue', strokeWidth=2).encode(
    x='Year:O',
    y='Percentile_50:Q'
)

# March 2025 forecast line
march_forecast = alt.Chart(gdp_for).mark_line(color='red', strokeWidth=2, strokeDash=[5,5]).encode(
    x='Year:O',
    y='March 2025 Forecast:Q'
)

# November 2025 forecast line
nov_forecast = alt.Chart(gdp_for).mark_line(color='green', strokeWidth=2).encode(
    x='Year:O',
    y='November 2025 Forecast:Q'
)

# Combine all layers
fan_chart = (fan_90_10 + fan_80_20 + fan_70_30 + fan_60_40 + central_line + march_forecast + nov_forecast).properties(
    width=700,
    height=400,
    title=alt.TitleParams(
        text='GDP Growth Forecast',
        fontSize=18
    )
)

fan_chart
```
