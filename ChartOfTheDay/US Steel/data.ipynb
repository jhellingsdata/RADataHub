```python
import pandas as pd
```

```python
steel = pd.read_excel('steel_data.xlsx')
```

```python
steel.columns = ['date', 'Mar 2024', 'Apr 2024', 'May 2024', 'Jun 2024', 'Jul 2024', 'Aug 2024', 'Sep 2024', 'Oct 2024', 'Nov 2024', 'Dec 2024', 'Jan 2025', 'Feb 2025', 'Mar 2025', 'Apr 2025']

steel
```

```python
# sort steel by the ' 2025-03-01 00:00:00' column
steel.sort_values(by='Mar 2025', ascending=False, inplace=True)

steel
```

```python
# delete the row where 'date' = "Total"
steel.drop(steel[steel['date'] == 'Total'].index, inplace=True)
steel
```

```python
# rename to 'country'
steel = steel.rename(columns={'date': 'country'})

# restrict to top 10 countries
steel = steel.head(7)

steel
```

```python
steel = steel.copy().melt(id_vars=['country'], var_name='date', value_name='value')

steel
```

```python
steel['date'] = pd.to_datetime(steel['date'], format='%b %Y')

steel.dtypes
```

```python
steel.to_csv('steel.csv', index=False)
```
