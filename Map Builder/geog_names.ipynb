### EU NUTS

```python
with open('eu_nuts2_topo.json', 'r') as f:
    topo_data = json.load(f)
```

```python
# Update the TopoJSON geometries
for geometry in topo_data['objects']['geog']['geometries']:
    if 'properties' in geometry and 'id' in geometry['properties']:
        nuts2_id = geometry['properties']['id']
        if nuts2_id in nuts2_names:
            # Create a new properties dictionary with 'id' and 'name' first
            new_properties = {
                'id': nuts2_id,
                'name': nuts2_names[nuts2_id]
            }
            # Add all other existing properties
            new_properties.update({k: v for k, v in geometry['properties'].items() if k != 'id'})
            # Replace the original properties with the new one
            geometry['properties'] = new_properties

# Save the updated TopoJSON
with open('eu_nuts2_topo.json', 'w', encoding='utf-8') as file:
    json.dump(topo_data, file, indent=2, ensure_ascii=False)
```
