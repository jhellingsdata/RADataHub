```python
fips_url = 'https://raw.githubusercontent.com/kjhealy/fips-codes/refs/heads/master/state_and_county_fips_master.csv'

fips = pd.read_csv(fips_url)
```

```python
# find only the rows in fips where the value in 'name' is all caps
fips[fips['name'].str.isupper()]
```

```python
states = fips[fips['name'].str.isupper()]
states
```

```python
# drop the first row of states
states = states.iloc[1:]
```

```python
# drop the last three digits of every number in the 'fips' column
states['fips'] = states['fips'].astype(str).str[:-3].astype(int)
states
```
