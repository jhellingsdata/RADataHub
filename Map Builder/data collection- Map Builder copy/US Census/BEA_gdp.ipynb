```python
import pandas as pd
import numpy as np
import json 
import time 
import requests
import pycountry
import csv
```

```python

url = 'https://apps.bea.gov/api/data/?UserID=01032369-1056-4225-A14B-44EF6C071679&method=GetData&datasetname=Regional&TableName=SAGDP9N&LineCode=2&Year=ALL&GeoFips=STATE&ResultFormat=json'

response = requests.get(url)

response
```

```python
data = response.json()
data
```

```python
# Extract the relevant data for columns: 'GeoName', 'TimePeriod', and 'DataValue'
data_list = data['BEAAPI']['Results']['Data']

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data_list)
df
```

```python
df2 = df[['GeoFips', 'GeoName', 'TimePeriod', 'DataValue']]

# Renaming the columns to 'code', 'name', 'date', and 'value'
df2 = df2.rename(columns={
    'GeoFips': 'id',
    'GeoName': 'name',
    'TimePeriod': 'date',
    'DataValue': 'value'
})

df2['date'] = df2['date'].astype(str) + '-01-01' 
df2 

df2.sample(20)
```

```python
df2.to_csv('gdp_usa_1997.csv', index=False)
```
