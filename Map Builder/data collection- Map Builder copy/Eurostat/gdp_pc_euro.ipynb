```python
# Dropping 'DATAFLOW' and 'LAST UPDATE' columns, and renaming 'TIME_PERIOD' to 'date'
df2 = df.drop(['DATAFLOW', 'LAST UPDATE', 'OBS_FLAG', 'freq', 'na_item'], axis=1).rename(columns={'TIME_PERIOD': 'date', 'OBS_VALUE': 'value'})

# Display the modified DataFrame
df2
```

```python
# FREQ = annual

# Filter the DataFrame where unit is "CLV10_EUR_HAB": "Chain linked volumes (2010), euro per capita"
df3 = df2.loc[(df2['unit'] == 'CLV10_EUR_HAB')]

# Display the filtered DataFrame
df3
```

```python
df3['date'] = df3['date'].astype(str) + '-01-01' 
df3
```

```python

# Define the function to convert geo codes to country names
def geo_to_country(geo_code):
    if geo_code == 'UK':
        return 'United Kingdom'
    if geo_code == 'EL':
        return 'Greece'
    country = pycountry.countries.get(alpha_2=geo_code)
    return country.name if country else 'Unknown'

# Example DataFrame

# Apply the function to add the new country column
df3['country'] = df3['geo'].apply(geo_to_country)

# Display the modified DataFrame
df3
```

```python
unknown_count = df3['country'].value_counts().get('Unknown', 0)

print(f"Count of 'Unknown' entries: {unknown_count}")

unknown_country_df = df3[df3['country'] == 'Unknown']

unknown_country_df
```

```python

# Function to get ISO3 code from country name
def get_iso3_code(country_name):
    try:
        country = pycountry.countries.lookup(country_name)
        return country.alpha_3
    except LookupError:
        return 'Unknown'

# Apply the function to the 'country' column and create the 'iso3' column
df3['iso3'] = df3['country'].apply(get_iso3_code)

# Display the modified DataFrame
df3
```

```python

df4 = df3.drop(['unit'], axis=1).rename(columns={'geo': 'id'})

df4
```

```python
df4.to_csv('gdp_pc_euro.csv', index=False)
```
