```python
import pandas as pd
```

```python
# download the data from the Census Bureau

states_pop = pd.read_csv('state_pops_23.csv')

# fix the strings in the 'state' column
states_pop['state'] = states_pop['state'].str.replace('.','')
states_pop.columns = ['name','pop']

states_pop
```

```python
import us

# get state fips codes and their names
state_fips = pd.DataFrame(us.states.mapping('fips', 'name'), index=[0]).T.reset_index()
# rename cols
state_fips.columns = ['fips', 'name']
# make fips an int
state_fips['fips'] = state_fips['fips'].astype(int)

state_fips
```

```python
# add the fips codes to the states_pop df
# states_pop = states_pop.merge(state_fips, on='name')

states_pop = states_pop[['fips','pop']]

# convert pop from string to int
states_pop['pop'] = states_pop['pop'].str.replace(',','').astype(int)

states_pop
```

```python
# output csv with no header
states_pop.to_csv('state_pops_23.csv', header=False, index=False)
```
