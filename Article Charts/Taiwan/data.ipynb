```python
import pandas as pd
```

```python
gdppc = pd.read_csv('gdppc.csv')

gdppc
```

```python
gdppc.columns
```

```python
# rename the first column 'Country'
gdppc.rename(columns={'GDP per capita, current prices (U.S. dollars per capita)': 'Country'}, inplace=True)

gdppc
```

```python
gdppc = pd.melt(gdppc, id_vars=['Country'], var_name='Year', value_name='GDP per capita')
gdppc
```

```python
gdppc['Country'].unique()
```

```python
gdp_sans_taiwan = gdppc[gdppc['Country'] != 'Taiwan Province of China']
gdp_taiwan = gdppc[gdppc['Country'] == 'Taiwan Province of China']
```

```python
# Convert 'GDP per capita' to numeric
gdp_sans_taiwan['GDP per capita'] = pd.to_numeric(gdp_sans_taiwan['GDP per capita'], errors='coerce')

# Group by 'Year' and calculate mean
gdp_avg = gdp_sans_taiwan.groupby('Year')['GDP per capita'].mean().reset_index()

# Add 'Country' column with value "G20 Average"
gdp_avg['Country'] = "G20 Average"

# Reorder columns to match the original DataFrame format
gdp_avg = gdp_avg[['Country', 'Year', 'GDP per capita']]

# Display the result
gdp_avg
```

```python
gdp_taiwan
```

```python
# concatenate the two DataFrames
taiwan_avg = pd.concat([gdp_avg, gdp_taiwan], ignore_index=True)
taiwan_avg
```

```python
# in the 'Country' column, replace 'Taiwan Province of China' with 'Taiwan'
taiwan_avg['Country'] = taiwan_avg['Country'].replace('Taiwan Province of China', 'Taiwan')
```

```python
# save to csv
taiwan_avg.to_csv('gdp_taiwan.csv', index=False)
```

```python
import requests
import pandas as pd

import matplotlib.pyplot as plt

# OEC API endpoint for bilateral trade data
base_url = "https://api.oec.world/olap-proxy/data"

# Parameters for Australia-Taiwan bilateral trade
params = {
    "cube": "trade_s_yp",
    "drilldowns": "Year",
    "measures": "Trade Value",
    "reporter": "aus",  # Australia
    "partner": "twn",   # Taiwan
    "year": "2013,2014,2015,2016,2017,2018,2019,2020,2021,2022",
}

# Make the API request
response = requests.get(base_url, params=params)
data = response.json()

# Process the data into a DataFrame
trade_data = pd.DataFrame(data['data'])

# Convert trade values to numeric and year to integer
trade_data['Trade Value'] = pd.to_numeric(trade_data['Trade Value'])
trade_data['Year'] = pd.to_numeric(trade_data['Year'])

# Plot the data
plt.figure(figsize=(12, 6))
plt.bar(trade_data['Year'], trade_data['Trade Value'] / 1e9)
plt.title('Bilateral Trade Between Australia and Taiwan')
plt.xlabel('Year')
plt.ylabel('Trade Value (Billion USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(trade_data['Year'])

# Add labels on top of each bar
for i, value in enumerate(trade_data['Trade Value']):
    plt.text(trade_data['Year'].iloc[i], value/1e9 + 0.2, 
             f"${value/1e9:.2f}B", 
             ha='center')

plt.tight_layout()
plt.show()

# Display the data
print("Australia-Taiwan Bilateral Trade Data:")
display(trade_data[['Year', 'Trade Value']])
```
