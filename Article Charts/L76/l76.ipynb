```python
import pandas as pd
import numpy as np
import eco_style
import altair as alt
alt.themes.enable('light')
```

```python
# The tab is "for_python"
# The x axis is "Area"
# The y-value for the three series are in columns E, F, G
# The size of the bubbles is column C or K (Recovery or duration)  they're the same thing

df = pd.read_excel("tent_collapses_crises_EO.xlsx", sheet_name="for_python")

# Country	Period	Recovery	Area	State Collapse / Decolonization	War / Civil War / Revolution	Neither	Min	start	end	duration
df = df.melt(
    id_vars=["Country", "Period", "Recovery", "Area", "Min","start", "end", "duration"],
    var_name="type",
).reset_index()


base = alt.Chart(df).add_selection(hover).encode(
    alt.X("Area:Q", axis=alt.Axis(labelAngle=-45), scale=alt.Scale(type="log", domain=[0.8,30])),
    alt.Y("value:Q", title="Peak-to-trough fall in GDP", 
          scale=alt.Scale(zero=False, domain=[-1, -0.3]),
          axis=alt.Axis(format="%")),
    alt.Size("Recovery:Q", scale=alt.Scale(range=[0.5, 800]), legend=alt.Legend(
        values=[50, 40, 30,20,10],
        titleColor="rgb(132,132,132)",
        labelExpr="datum.value == 50 ? '50 years' : datum.value",
        legendX=2,
        legendY=-75,
        title="Recovery time",
    )),
    color=alt.Color("type:N", legend=alt.Legend(title="Type of crisis",
                                                    values=["War / Civil War / Revolution", "State Collapse / Decolonization", "Neither"],
                                                    titleColor="rgb(132,132,132)",
                                                    ),
    ),
    tooltip=["Country", "start", "end", "duration", "type"],
)

bubbles = base.mark_circle(
    opacity=alt.expr("datum.Country == 'Venezuela' ? 1 : 0.5"),
).encode(
)

venuz_label = base.transform_filter(
    'datum.Country == "Venezuela"'
).mark_text(
    align="left",
    baseline="middle",
    dx=7,
    dy=0,
    fontSize=10,
    color="black",
    text=alt.expr("datum.Country + ' (' + datum.Period + ')'"),
).encode(
    size=alt.value(10),
)

chart = (bubbles + venuz_label).interactive().properties(
    width=450,
    height=350
)

chart.save("l76.json")
    

chart.interactive()
```
