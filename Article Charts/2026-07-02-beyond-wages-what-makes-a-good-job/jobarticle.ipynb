```python
import pandas as pd
import numpy as np
import altair as alt

# Using altair theme from local eco_style file
# import eco_style 
# alt.themes.enable('report')  


# Using ecostyles package (installed in environment)
from ecostyles import EcoStyles
styles = EcoStyles()
styles.register_and_enable_theme(theme_name='article')

from pathlib import Path
print(Path.cwd())
```

```python
# Format article title: remove any special characters, replace spaces with hyphens, and convert to lowercase
title = "Beyond wages: what makes a good job?"
post_date = '2026-07-02'
title_fmt = post_date + '-' + title.replace("?", "").replace(":", "").replace(" ", "-").lower()
title_fmt
```

```python
jobchart = pd.read_csv("jobchartdata.csv")
worktime = pd.read_csv("workingtimedata.csv")
```

```python
#pivoting jobchart
jobchart["id"] = jobchart["id"].str.strip()

jobchart_category = jobchart.pivot(
    index="category",
    columns="id",
    values="value"
).reset_index()

print(jobchart_category)
```

```python
bars = alt.Chart(jobchart_category).mark_bar().encode(
    x=alt.X('primary:Q', title=None),
    y=alt.Y('category:N', sort='-x', title=None),
    tooltip=[
        'category',
        alt.Tooltip('primary:Q', title='Estimate'),
        alt.Tooltip('low:Q', title='Low'),
        alt.Tooltip('high:Q', title='High')
    ]
)

error_bars = alt.Chart(jobchart_category).mark_rule(size=1, color='#122b39').encode(
    x='low:Q',
    x2='high:Q',
    y=alt.Y('category:N', sort='-x'),
    tooltip=[alt.Tooltip('low:Q', title='Low'), alt.Tooltip('high:Q', title='High')]
)

ticks = alt.Chart(jobchart_category).mark_tick(thickness=1, size=10, color='#122b39').encode(
    x='low:Q', y=alt.Y('category:N', sort='-x')
) + alt.Chart(jobchart_category).mark_tick(thickness=1, size=10, color='#122b39').encode(
    x='high:Q', y=alt.Y('category:N', sort='-x')
)

chart = alt.layer(bars, error_bars,ticks).properties(
    width=420,
    height=300
).resolve_scale(
    y='shared'
)


styles.save(chart, 'charts', 'fig1_job_quality', width=420, height=300)

```

```python
#zero line not emphasized
chart1 = alt.Chart(worktime).mark_line(point=True).encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('value:Q', title='Working time quality',scale=alt.Scale(domain=[-0.2, .2])),
    color=alt.Color('gender:N', title='Gender'),
    tooltip=['year', 'gender', 'value']
).properties(
    width=420,
    height=300
)

styles.save(chart1, 'charts', 'fig2_working_time_quality', width=420, height=300)
```
