```python
from ecostyles import EcoStyles
# Create styles instance
styles = EcoStyles()

import altair as alt
import pandas as pd
```

```python
# Register and enable a theme
styles.register_and_enable_theme(theme_name="article")
```

```python
# Load Excel file as df
wc_df = pd.read_excel('wc2026-probs.xlsx')

# Reset column index
wc_df = wc_df.rename(columns={'Unnamed: 0': 'id'}).set_index('id')

# Rename to capitalize country names
wc_df['team'] = wc_df['team'].str.title()

# Fill NaN probability values with 0
prob_cols = ['p.l32', 'p.l16', 'p.qf', 'p.sf', 'p.f', 'p.w']
wc_df[prob_cols] = wc_df[prob_cols].fillna(0)

wc_df
```

```python
# Create top 15 for bar chart
top15_df = wc_df.nlargest(15, 'p.w').reset_index()
top15_df
```

```python
# Create top 10 for slope chart
top10_df = wc_df.nlargest(10, 'p.w').reset_index()

# Specify stage order
stage_order = ['p.l32', 'p.l16', 'p.qf', 'p.sf', 'p.f', 'p.w']

# Melt top 10 into long format for charting
slope_long = top10_df.melt(
    id_vars=['team'],
    value_vars=stage_order,
    var_name='stage',
    value_name='probability'
)

# Make stage ordered for correct x-axis sequence
slope_long['stage'] = pd.Categorical(slope_long['stage'], categories=stage_order, ordered=True)
```

```python
# Create winner probability bar chart
win_bar = alt.Chart(top15_df).mark_bar().encode(
    x = alt.X(
        'p.w:Q',
        title='Probability of winning (%)',
        axis=alt.Axis(grid=False)),
    y = alt.Y(
        'team:N',
        sort=alt.EncodingSortField(field='p.w', op='sum', order='descending'),
        title=''),
    tooltip = [
        alt.Tooltip('team:N', title='Team'),
        alt.Tooltip('p.w:Q', title='Win probability (%)', format='.2f')
    ]
)
win_bar
```
