```python
from ecostyles import EcoStyles
# Create styles instance
styles = EcoStyles()
# Register and enable a theme
styles.register_and_enable_theme(theme_name="article")  # or "article"
```

```python
net_migra = pd.read_csv('PEA15.20251025T151051.csv')
```

```python
net_migra['Year'] = pd.to_datetime(net_migra['Year'], format='%Y')
```

```python
net_migra['VALUE_label'] = [f'{i}K' for i in net_migra['VALUE']]
```

```python
line = alt.Chart(net_migra).mark_line().encode(
    x=alt.X('Year', axis=alt.Axis(labelFontSize=14)),
    y=alt.Y('VALUE:Q', scale=alt.Scale(domain=[-100, 120]), axis=alt.Axis(labelExpr="datum.value + 'K'", labelFontSize=14)),
    tooltip=[
        alt.Tooltip('Year:T', format = '%Y', title='Year'),
        alt.Tooltip('VALUE_label', title='Net Migration')
    ],
    color=alt.value('#2a9d8f')
).properties(
    width=600,
    height=400)

zero_line = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(strokeWidth=2, color='lightgrey').encode(
    y='y:Q'
)

chart = zero_line +line
chart
```

```python
# Save to png
chart.save('ireland_net_migration.png', scale_factor=2)
# Save to json
chart.save('ireland_net_migration.json', scale_factor=2)
```
