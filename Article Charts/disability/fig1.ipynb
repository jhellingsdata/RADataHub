```python
import eco_style
import pandas as pd
import altair as alt
import numpy as np
alt.themes.enable('light')
```

```python
df = pd.read_csv("fig1.csv")
df["date"] = df.period.str[:4]+df.period.str[-4:]
df['date'] = pd.to_datetime(df['date'], format='%b-%Y')
df['value'] = pd.to_numeric(df['value'], errors='coerce')/100

base = alt.Chart(df).encode(
    x=alt.X('date:T', title=''),
    y=alt.Y('value:Q', title='', axis=alt.Axis(format='%')),
    detail=alt.Detail('series:N'),
)

lines = base.mark_line(color="#36b7b4")

# rule and label for the reweighting of the LFS in 2019
reweight_df = pd.DataFrame({
    'date': [pd.to_datetime('2019-01-01')],
    'label': ['Labour Force Survey\nReweighted'],
})

reweight_df['label'] = reweight_df['label'].apply(lambda x: x.split('\n'))

rule = alt.Chart(reweight_df).mark_rule(color='black', strokeDash=[5, 3], opacity=0.4).encode(
    x='date:T',
)

rule_text = rule.mark_text(
    align='left',
    baseline='middle',
    dx=3,
    dy=-10,
    fontWeight='normal',
    color='#676A86',
    opacity=0.7,
).encode(
    x='date:T',
    text=alt.Text('label:N', format=''),
    y=alt.value(0)
)

chart = (lines+rule+rule_text).properties(
    width=500,
    height=350
)

chart.save('fig1.json')
chart.save('fig1.png', scale_factor=2)
chart.save('fig1.svg')
chart
```
