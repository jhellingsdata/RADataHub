```python
import pandas as pd
import numpy as np
import altair as alt
import eco_style 
alt.themes.enable('light')    
```

# Fig1

```python

df = pd.read_csv("intermediate/ge_1924_2024_agg.csv")
#df['year'] = pd.to_numeric(df['year']).astype(str)

df = df[['year', 'Labour_Leader', 'Labour_Seats', 'Total_Seats']]
df['Labour_share'] = df['Labour_Seats'] / df['Total_Seats']

df.iloc[12, 0] = '1974F'
df.iloc[13, 0] = '1974O'

df = df.sort_values('Labour_share', ascending=False)
df['rank'] = np.arange(1, len(df)+1)

base = alt.Chart(df).encode(
 y=alt.Y('year:N', title='', sort=alt.EncodingSortField(field='rank', order='ascending')),
 x=alt.X('Labour_share', title='', axis=alt.Axis(format='%')),
 opacity=alt.condition(alt.datum.year == 2024, alt.value(1), alt.value(0.7))
)

bars = base.mark_bar(color="rgb(203, 43, 56)")

leader_labels = base.mark_text(
    align='left',
    baseline='middle',
    color='#676A86',
    font='Circular Std',
    fontSize=9,
    dx=3
).encode(
    text='Labour_Leader',
    color=alt.condition(alt.datum.year == 2024, alt.value('rgb(203, 43, 56)'), alt.value('#676A86'))
)

# add a rule at x=0.632308

rule = alt.Chart(pd.DataFrame({'x': [0.632308]})).mark_rule(color='#676A86', strokeDash=[3,5], opacity=0.5).encode(x='x')

chart = (bars + rule +leader_labels)

chart.save("fig1.json")
chart.save("fig1.png", scale_factor=2.0)

chart
```

# Fig 3

```python
ge_1924_2024_agg = pd.read_csv("intermediate/ge_1924_2024_agg.csv")

# Calculate vote shares and seat shares
ge_1924_2024_agg['Conservative_Party_VS'] = ge_1924_2024_agg['Conservative_Votes'] / ge_1924_2024_agg['Total_Votes']
ge_1924_2024_agg['Labour_Party_VS'] = ge_1924_2024_agg['Labour_Votes'] / ge_1924_2024_agg['Total_Votes']
ge_1924_2024_agg['Conservative_Share_Seats'] = ge_1924_2024_agg['Conservative_Seats'] / ge_1924_2024_agg['Total_Seats']
ge_1924_2024_agg['Labour_Share_Seats'] = ge_1924_2024_agg['Labour_Seats'] / ge_1924_2024_agg['Total_Seats']

# Select relevant columns for Labour Party
t2 = ge_1924_2024_agg[['Labour_Leader_VS', 'Labour_Leader_VS_LastElection',
                       'Labour_Party_VS', 'Labour_Share_Seats', 'Labour_Leader',
                       'GE_Year']].copy()

# Create Labour_Leader_Year column
t2['Labour_Leader_Year'] = t2['Labour_Leader'] + '; ' + t2['GE_Year'].astype(str)

# Rename columns
t2.columns = t2.columns.str.replace('Labour_', '')

# Add Party column
t2['Party'] = 'Labour'

# Calculate changes
t2['Change_Party_VS'] = t2['Party_VS'] - t2['Party_VS'].shift()
t2['Change_Share_Seats'] = t2['Share_Seats'] - t2['Share_Seats'].shift()
t2['Change_Leader_VS'] = t2['Leader_VS'] - t2['Leader_VS_LastElection']

# Create Leader_Year1 column
t2['Leader_Year1'] = t2['Leader_Year'].str.replace(r'^[^ ]+ ', '', regex=True)

# Add specific labels
t2.loc[t2['GE_Year'] == '2024', 'LabelText'] = 'Starmer; 2024'
t2.loc[t2['GE_Year'] == '1997', 'LabelText'] = 'Blair; 1997'

t2['LabelText'] = t2['LabelText'].str.replace("nan", "")

```

```python
base = alt.Chart(t2).encode(
    x=alt.X('Change_Leader_VS', title="Change in leader's own vote share", axis=alt.Axis(
        titleX=105,
        format='%')),
    y=alt.Y('Change_Share_Seats', title='Change in Labour seat share', axis=alt.Axis(format='%')),
    color=alt.value('rgb(203, 43, 56)'),
)

points = base.mark_circle()

labels = base.mark_text(
    align='left',
    baseline='middle',
    color='#676A86',
    font='Circular Std',
    fontSize=9,
    dx=3
).encode(
    text='LabelText',
)

# add x=0 and y=0 lines
rule1 = alt.Chart(pd.DataFrame({'x': [0]})).mark_rule(color='#676A86', strokeDash=[3,3], opacity=0.5).encode(x='x')
rule2 = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='#676A86', strokeDash=[3,3], opacity=0.5).encode(y='y')

chart = rule1+rule2+points + labels

chart.save("fig3.json")
chart.save("fig3.png", scale_factor=2.0)

chart
```

# Fig 4

```python
ge_2010_2024 = pd.read_csv("intermediate/ge_2010_2024.csv")

t1 = ge_2010_2024[
    (ge_2010_2024['CandidateName'].isin(["Keir Starmer", "Angela Rayner",
                                         "Rachel Reeves", "Yvette Cooper",
                                         "David Lammy", "Wes Streeting", 
                                         "Bridget Phillipson"])) &
    (ge_2010_2024['GE_ElectionYear'] >= 2019)
].copy()

df = t1.copy()
df = df[['CandidateName', 'GE_ElectionYear', 'VoteShare_LAB']]
df

base = alt.Chart(df).encode(
    row = alt.Row("CandidateName:N", header=alt.Header(title="", labelAngle=0, labelAlign='left' )),
    x=alt.X("VoteShare_LAB", axis=alt.Axis(format='%'), title=''),
    y=alt.Y("GE_ElectionYear:N", axis=alt.Axis(domain=False, ticks=False, labels=False, title='')),
)

bars = base.mark_bar(color="rgb(203, 43, 56)").encode(
    opacity=alt.Opacity("GE_ElectionYear:N", 
                        legend=alt.Legend(
                            orient='none',
                            direction='horizontal',
                            title='',
                            legendY=-15
                        ),
                        scale=alt.Scale(domain=[2019, 2024], range=[1, 0.5]))
)

chart = bars

chart = chart.properties(
    height=40
)

chart.save("fig4.json")
chart.save("fig4.png", scale_factor=2.0)

chart
```

# Fig 6

```python
ge_2024 = pd.read_csv("intermediate/ge_2024.csv")
ge_2024['Seat_Labour'] = ge_2024['PartyNameWinner'].apply(lambda x: 1 if x == 'LAB' else 0)
ge_2024['Seat_Conservative'] = ge_2024['PartyNameWinner'].apply(lambda x: 1 if x == 'CON' else 0)
ge_2024['Seat_LibDem'] = ge_2024['PartyNameWinner'].apply(lambda x: 1 if x == 'LIBDEM' else 0)
ge_2024['Seat_Other'] = ge_2024['PartyNameWinner'].apply(lambda x: 1 if x not in ['LAB', 'CON', 'LIBDEM'] else 0)

# Calculate mean seat shares for different incumbent scenarios
t1 = ge_2024[(ge_2024['IncumbentLabourCand'] == 1) & (ge_2024['IncumbentLabour'] == 1)][[
    'Seat_Conservative', 'Seat_Labour', 'Seat_Other', 'Seat_LibDem'
]].mean().reset_index()
t1.columns = ['variable', 'value']
t1['Type'] = "Labour\nIncumbent Candidate\nRe-Contesting"
t1['descrip'] = ["Conservative Gain", "Labour Retain", "Other Gain", "Liberal Democrats Gain"]

t2 = ge_2024[(ge_2024['IncumbentLabourCand'] == 0) & (ge_2024['IncumbentLabour'] == 1)][[
    'Seat_Conservative', 'Seat_Labour', 'Seat_Other', 'Seat_LibDem'
]].mean().reset_index()
t2.columns = ['variable', 'value']
t2['Type'] = "Labour\nIncumbent Candidate\nNot Re-Contesting"
t2['descrip'] = ["Conservative Gain", "Labour Retain", "Other Gain", "Liberal Democrats Gain"]

t3 = ge_2024[(ge_2024['IncumbentConservativeCand'] == 1) & (ge_2024['IncumbentConservative'] == 1)][[
    'Seat_Conservative', 'Seat_Labour', 'Seat_Other', 'Seat_LibDem'
]].mean().reset_index()
t3.columns = ['variable', 'value']
t3['Type'] = "Conservative\nIncumbent Candidate\nRe-Contesting"
t3['descrip'] = ["Conservative Retain", "Labour Gain", "Other Gain", "Liberal Democrats Gain"]

t4 = ge_2024[(ge_2024['IncumbentConservativeCand'] == 0) & (ge_2024['IncumbentConservative'] == 1)][[
    'Seat_Conservative', 'Seat_Labour', 'Seat_Other', 'Seat_LibDem'
]].mean().reset_index()
t4.columns = ['variable', 'value']
t4['Type'] = "Conservative\nIncumbent Candidate\nNot Re-Contesting"
t4['descrip'] = ["Conservative Retain", "Labour Gain", "Other Gain", "Liberal Democrats Gain"]

# Combine data
t5 = pd.concat([t1, t2, t3, t4], ignore_index=True)

# Convert values to percentages
t5['value'] = (t5['value'] * 100).round(0)

# Add party colors
t5['party_color'] = t5['variable'].apply(lambda x: "#0073e6" if "Conservative" in x else
                                                       "red" if "Labour" in x else
                                                       "gray" if "Other" in x else
                                                       "orange")

# Custom labels for the legend
legend_labels = {
    "Seat_Conservative": "Conservative",
    "Seat_Labour": "Labour",
    "Seat_Other": "Others",
    "Seat_LibDem": "Liberal Democrats"
}

# Add description for labels
t5['description'] = t5.apply(lambda row: f"{row['descrip']}: {row['value']}%", axis=1)

t5['value'] = t5['value']/100
###############


rows = []

for i, var_type in enumerate(t5['Type'].unique()):
    temp_df = t5[t5['Type'] == var_type]

    base = alt.Chart(temp_df).encode(
        y=alt.Y('variable:N', title='', axis=alt.Axis(domain=False, ticks=False, labels=False, title='')),
        x=alt.X('value:Q', title='', 
                scale=alt.Scale(domain=[0,1]),
                axis=((alt.Axis(format='%')) if i==3 else alt.Axis(domain=False, ticks=False, labels=False, title=''))
                
        ),
        color=alt.Color('variable:N', legend=None,
                        scale=alt.Scale(
                            domain=[
                                'Seat_Labour',  'Seat_Conservative', 'Seat_LibDem', 'Seat_Other'
                            ],
                            range=['rgb(203, 43, 56)', 'rgb(43, 115, 203)', 'rgb(203, 131, 43)', 'rgb(172, 172, 172)']
                        )
                        ),
        tooltip='description'
    )

    bars = base.mark_bar()

    labels = base.mark_text(
        align='left',
        dx=alt.expr("datum.value == 0 ? 0 : 5")
    ).encode(
        text=alt.Text("description")
    )

    chart = bars + labels

    chart = chart.properties(
        height=75,
        title=alt.TitleParams(
            text=var_type.split("\n"),
            orient='none',
            align='right',
            anchor='end',
            dy=10,
            dx=-15,
            fontWeight='normal',
            fontSize=10,
            
            angle=0
        )
    
    )

    rows.append(chart)

chart = alt.vconcat(*rows).resolve_axis(x="shared")

chart.save("fig6.json")

chart
    
```
