```python
df = pd.read_csv("gfcf.csv")
df = df.melt(id_vars=["Year"], var_name="Country", value_name="Value")
df['Date'] = pd.to_datetime(df['Year'], format='%Y')
df['Value'] = df['Value'].str.replace("%", "").astype(float)/100
df
```

```python
alt.themes.enable('dark')

base = alt.Chart(df).encode(
    x=alt.X('Date:T', axis=alt.Axis()),
    y=alt.Y('Value:Q', axis=alt.Axis(title='Gross Fixed Capital Formation (% of GDP)', format='%') ),
    color=alt.Color('Country:N', legend=None)
)

line = base.mark_line()

labels = base.transform_filter(
    "year(datum.Date) == 2023"
).mark_text(
    align='left',
    baseline='middle',
    dx=3,
    dy=alt.expr("datum.Country == 'US' ? 5  : 0"),
).encode(
    text=alt.Text('Country:N')
)

chart = line + labels

chart.save("gfcf_country.json")
chart.save("gfcf_country.png", scale_factor=2.0)

chart 
```

## Chart 2

```python
df = pd.read_csv("ratios.csv")
df = df.melt(id_vars=["Date"], var_name="Series", value_name="Value")
# Quarters to datres
df['Date'] = pd.to_datetime(df['Date'].str[:4] + "-" + (((df["Date"].str[-1].astype(int))-1)*3+1).astype(str), format='%Y-%m')
df = df.reset_index()
df
```

```python
alt.themes.enable('report')

base = alt.Chart(df).encode(
    x=alt.X('Date:T', axis=alt.Axis()),
    y=alt.Y('Value:Q', axis=alt.Axis(title='Shares of consumption, investment and net trade', format='%'), scale=alt.Scale(domain=[-0.06, 0.06]) ),
    color=alt.Color('Series:N', legend=None, type='nominal')
)

line = base.mark_line(interpolate='monotone')

labels = base.transform_filter(
    "year(datum.Date) == 2023 & month(datum.Date) == 0"
).mark_text(
    align='left',
    baseline='middle',
    dx=3,
    dy=alt.expr("datum.Series == 'Share of Investments' ? -15  : 0"),
).encode(
    text=alt.Text('Series:N')
)

rule = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(
    color="#676A86",
    opacity=0.5,
).encode(y='y')


chart = rule + line + labels

chart.save("gfcf_ratios.png", scale_factor=2.0)
chart.save("gfcf_ratios.json")

chart.display()
```
