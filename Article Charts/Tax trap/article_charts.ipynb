```python
df1 = pd.read_csv('buching_figure.csv')
```

```python
df1 = df1[(df1['bin'] >= 75000) & (df1['bin'] <= 125000)]

line = alt.Chart(df1).mark_line().encode(
    x=alt.X('bin', 
            scale=alt.Scale(domain=[75000,125000]), 
            axis=alt.Axis(title="",#"Annual Income",
                          titleFontSize=14, 
                          labelExpr="'£' + datum.value / 1000 + 'K'",
                          labelFontSize=13, 
                          tickCount=6)),
    y=alt.Y('weighted_count', 
            scale=alt.Scale(domain=[10000,70000]), 
            axis=alt.Axis(title="",#"Taxpayers",
                          titleFontSize=14,
                          titleX= -29,
                          titleY = -10, 
                          grid=False, 
                          labelExpr="datum.value / 1000 + 'K'",
                          labelFontSize=13,
                          )),
    tooltip=[alt.Tooltip('weighted_count', format=',.0f', title='Taxpayers'),
            alt.Tooltip('bin', format=',.0f', title='Income (£)')]
)

vertical_line = alt.Chart(pd.DataFrame({'x': [100000]})).mark_rule(color='black', strokeDash=[5,5]).encode(
    x='x'
)

chart = line + vertical_line

chart.properties(
    width=500,
    height=300,
    title=alt.TitleParams(
        text="",#'Taxpayers Limit Income to Avoid the "Tax Trap"',
        fontSize=16,
        anchor="start"))
```

```python
df2 = pd.read_csv('marginal_tax_rates.csv')
```

```python
df2['type'].unique()
```

```python
typemap = {
    'income_tax': 'Income Tax',
    'plus_nics': '+ National Insurance',
    'plus_PA_withdrawl': '+ Personal Allowance Withdrawal',
    'plus_child_benefit_withdrawal': '+ Child Benefit Withdrawal',
    'plus_student_loans': '+ Student Loans'
}
df2['type'] = df2['type'].map(typemap)
```

Figure 1

```python
alt.Chart(df2).mark_line().encode(
    x=alt.X('income', axis=alt.Axis(title="",#"Annual Income",
                          titleFontSize=14, 
                          labelExpr="'£' + datum.value / 1000 + 'K'",
                          labelFontSize=13, 
                          tickCount=6,
                          grid=False)),
    y=alt.Y('rate', axis=alt.Axis(labelFontSize=14, format='%')),
    color=alt.Color('type', legend=alt.Legend(orient='bottom'))
).properties(
    width=600,
    height=300)
```
