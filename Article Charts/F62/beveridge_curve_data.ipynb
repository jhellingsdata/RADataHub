```python
import pandas as pd
```

```python
bev = pd.read_excel('Bevridge_Curves_UK_Data.xlsx', sheet_name='Sheet1')
bev
```

```python
# rename cols
bev.columns = ['date', 'unemployment', 'vacancy']

# convert date from object like 2014 JAN to datetime
bev['date'] = pd.to_datetime(bev['date'], format='%Y %b')

bev
```

```python
# show only rows where date is in 2021
bev[bev['date'].dt.year == 2021]
```

```python
bev[bev['date'].dt.year == 2022]
```

```python
# create a 'Time Period' string column: the values should be '2014-2019', '2022', '2023', and '2024'

# create an empty string column
bev['Time Period'] = ''

# for march 2014 to feb 2019, set the Time Period to '2014-2019'
bev.loc[(bev['date'] >= '2014-03-01') & (bev['date'] <= '2019-02-28'), 'Time Period'] = '2014-2019'

# for march 2022 to feb 2023, set the Time Period to '2022'
bev.loc[(bev['date'] >= '2022-03-01') & (bev['date'] <= '2023-02-28'), 'Time Period'] = '2022'

# for march 2023 to feb 2024, set the Time Period to '2023'
bev.loc[(bev['date'] >= '2023-03-01') & (bev['date'] <= '2024-02-29'), 'Time Period'] = '2023'

# for march 2024 to feb 2025, set the Time Period to '2024'
bev.loc[(bev['date'] >= '2024-03-01'), 'Time Period'] = '2024'

# if the Time Period is still empty, set it to a missing value
bev.loc[bev['Time Period'] == '', 'Time Period'] = pd.NA
```

```python
bev.to_csv('beveridge_curve_data.csv', index=False)
```
